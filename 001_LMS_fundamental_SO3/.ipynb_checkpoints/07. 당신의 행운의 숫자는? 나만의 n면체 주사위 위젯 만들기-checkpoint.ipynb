### 변수

 - `myvar = 3`
     - `=` 연산자를 이용해 값을 할당하는 것
     - 복사하는 것이 아니라, 데이터가 담긴 객체에 **그냥 이름을 붙이는 것**이다.
     - **참조**할 수 있게 하는 것
 - `myvar` : 3을 담은 객체
     - `3` : 의미적으로는 `3이라는 고유한 값.`이라고 이야기할 수 있겠지만, 정확하게는 `3` 또한 객체이다.
     - 즉 `3`은 정수 타입의 객체이다.
 - `myword = 'cat'`
     - `cat`이라는 문자열 타입의 객체를 생성하고, `myword`라는 변수(객체)에 할당하여 문자열 객체를 참조하게 한다.
 - `myword.upper()` / `cat.upper()`
     - 객체 `myword`/`cat`의 매서드
 - `id()`
     - 파이썬의 내장함수(`bulletin_function`)
     - 고윳값(identity)을 반환한다. 고윳값이란
         - 객체마다 부여된 유일한 값 혹은
         - 파이썬에서는 '메모리주소'라고 이해하면 된다.
     - 변수를 할당하고 (`var = 4`) `var`과 `4`의 고윳값을 각각 계산하면 같은 결과가 나온다.
     이것은 `is` 명령어를 통해서도 확인할 수 있을 것이다.

```python
var = 4
print(type(id))
print(id(4))
print(id(var))
```

### 얕은 복사, 깊은 복사

얕은복사와 깊은 복사에 대해서는 어제인가 그제, 확실히 이해했었다.
어쩌면 확실히 이해했다고 생각했을지도 모르겠지만, 여하튼 그 순간에는 이해했었다.

그리고 `python` 공식 홈페이지의 `copy` documentation도 어느 정도는 술술 읽힌다.
고 생각했다.
몇몇 문장들은 여기에 옮겨적어 보겠다.
이 문장들은 정확하게 이해하고 있다. (고 생각한다.)

Assignment statements in Python do not copy objects, they create bindings between a target and an object.
For collections that are mutable or contain mutable items, a copy is sometimes needed so one can change one copy without changing the other.

The difference between shallow and deep copying is only relevant for compound objects (objects that contain other objects, like lists or class instances):
 - A shallow copy constructs a new compound object and then (to the extent possible) inserts references into it to the objects found in the original.
 - A deep copy constructs a new compound object and then, recursively, inserts copies into it of the objects found in the original.


그런데 생각만큼 실제 코드는 잘 이해되지 않는다.
연산자 `==`의 동작은 잘 이해된다.
하지만 파이썬 내장함수 `id`의 동작은 이해하기 어렵다.
마찬가지의 맥락에서 연산자 `is`도 내가 이해한 바와 맞아떨어지지 않는다.

이 부분은 나중에 시간이 지나서 다시 보면 좋겠다.
또한, 나중에 `list`, `tuple`, `dict`, `set`에 대하여 한번에 정리해보는 것도 좋겠다.

참고로, 이미 정리해놓은 것은 `dict`에 대한 copy-deepcopy이다. 코딩도장 25.5에 있다.
(두 번 정도 정리했던 것으로 기억하는데, 일단 확인되는건 `dict` 뿐이다.)

```python
import copy
A = [[1,2],[3,4]]
# A = {'first':{'a':1,'b':2},'second':{'c':3,'d':4}}
B = A # 할당 (assign)
C = A.copy() # 얕은 복사 (copy)
D = copy.deepcopy(A) #깊은 복사 (deep copy)

print('A :',A)
print('B :',B)
print('C :',C)
print('D :',D)
print()
print('B == A :',B == A)
print('C == A :',C == A)
print('D == A :',D == A)
print()
print('id(A) :',id(A))
print('id(B) :',id(A))
print('id(C) :',id(A))
print('id(D) :',id(A))
print()
print('B is A :',B is A)
print('C is A :',B is A)
print('D is A :',B is A)
```

### 요약
> 파이썬에서는 모든 것(부울, 정수, 실수, 데이터구조(`list`,`tuple`,`dict`,`set`...), 함수, 프로그램, 모듈)이 객체다.
객체는 상태(state)를 나타내는 *속성(attribute)*과 동작(behavior)을 나타내는 *메서드(method)*가 있다.
객체의 속성은 변수로 구현된다.
객체의 메서드는 함수로 구현된다.

## 7.3. 객체 지향 프로그래밍 (2) 객체 지향 프로그래밍이란?

 - **객체 지향 프로그래밍** (OOP : Object Oriented Programming)
 - '객체만' 이용한 프로그래밍이 부각됐다.
   - 이것은 수학에서 집합론의 관점을 보는 것 같다.
   Halmos의 「Naive Set Theory」에서 '모든 것은 집합이며, 그 이외의 것들은 생각할 필요가 없다. 조금 더 정확하게 말하면, 모든 것들이 집합이라고 가정해도 아무런 논리적인 비약이 생기지 않기 때문에 그렇게 한다'라고 했던 것 같다.
   즉 집합에 표현된 각각의 원소조차도, 하나의 집합이라고 간주해도 문제가 없다는 것이다.
 - 객체 지향적인 관점에서 보면 **변수의 묶음**이 중요하게 부각된다.
 - 예시 : 도서관 관리프로그램
 - 변수를 함수로 어떻게 처리해야 하는가? --> 절차 지향 프로그래밍
 - **절차 지향 프로그래밍** (Procedural Programming)에서는 '변수의 묶음'에는 관심이 없고 **처리**에 관심이 있다.
 효율적이기 때문에 나름의 장점이 있다.
 - 위의 두 형태는 서로 상반되는 프로그래밍 방식이기는 하지만 완전히 서로 배척되는 것은 아니다.
 또한, 위의 두 형태 말고도 다른 종류의 프로그래밍 방식도 존재한다.
 - 다 다루기에는 너무 양이 많기 때문에, 이번 노드에서는 **원리가 아닌 실습** 위주로 설명할 예정이다.

## 7.4. 클래스 기본 문법 (1) 클래스 선언 및 인스턴스화

객체 설계의 가장 기본 개념 : `class`

### 클래스 선언

```python
class Car:
    pass

class Car():
    pass

print(Car)
print(Car())

# 두 객체의 type을 살펴봅니다. 
print(type(Car))
print(type(Car()))

#id(Car())는 Car()가 호출될 때마다 다른 값이 얻어집니다. 
print(id(Car()))
print(id(Car()))

# 두 객체의 type을 살펴봅니다. 
print(type(Car))
print(type(Car()))
```

```python
print(type(list))
print(type(tuple))
print(type(dict))
print(type(set))
print(type(str))
```

 - 맨 앞 다섯 줄에서 `class`를 두 번 정의하고 있다.
    - <font color=red>이 정의는 같은 정의인가?</font>
    - 또한, 같은 이름으로 두 번 정의했으니, 위의 선언은 무시되는 것인가?
    - 아니다, 아래 내용 상 서로 다른 두 개의 `class`가 정의된 것이라고 봐야 할 것 같다.
 - `type`을 먼저 보자.
  - `Car`는 클래스 그 자체이다.
      - 그리고 클래스는, `type`했을 때 `<Class : type>`이라고 표시된다.
      - 밑에 그 예시를 적어놓았는데, `list`도 하나의 클래스이다.
      따라서 `type`했을 때 마찬가지로 `<Class : type>`이라고 표시된다.
  - `Car()`는 `Car` 클래스로 만든 객체이다.
 - `id`가 출력하는 결과를 보자.
   - `Car`는 여러 번 호출되더라도 그 `id`값이 항상 같다.
   - `Car`의 객체인 `Car()`는 호출될 때마다 그 `id`값이 달라질 수 있다.

### 클래스 사용 - 객체 인스턴스화
 - **인스턴스화** : 클래스를 사용해 객체를 만드는 것
   - 클래스명 다음에 소괄호를 열었다가 닫으면 된다.
   - `Car()`
   - "클래스를 호출했다."라는 표현을 사용한다.

### 표기법
 - 클래스의 개념은, 클래스명 다음에 소괄호를 열었다가 닫으며, 그 사이에 인자를 넣을 수 있다는 점에서 함수와 매우 유사하다.
 심지어는 헷갈릴 여지가 있다.
 - 그래서 클래스와 함수를 표기하는 데 있어 특정한 convention(관습)이 있다.
 - 이와 같은 관습 혹은 권장사항 중 표준이 되는 것은 'PEP이다.
   - Python Enhancement Proposals
   - PEP8
 - 함수와 클래스의 표기법 권장사항
    - 클래스는 PascalCase$*$로 적는다.
    - 함수는 snake_case로 적는다.
    - 클래스는 명사에서 주로 차용한다.
    - 함수는 동사에서 주로 차용한다.
    
$*$ 교재에 오류가 좀 있는 것 같다. camelCase가 아니라 PascalCase가 맞지 않을까 싶다. 여기에는 오류를 정정하여 옮겼다.

## 7.5. 클래스 기본 문법 (2) 클래스 속성과 메서드

 - 클래스는 객체를 설계하기 위한 도구.
 - *속성(attribute)* : 상태(state)를 표현한다. ---> *변수(variable)*
 - *메서드(method)* : 동작(behavior)를 표현한다. ---> *함수(function)*
 |구분|*속성(attribute)*|*메서드(method)*|
 |:-:|:-:|:-:|
 |의미|상태(state)를 표현한다.|동작(behavior)을 표현한다.|
 |일반 python에서 대응되는 것|*변수(variable)*|*함수(function)*|

예시 : `Car` 클래스 코딩하기.
 - 손으로 코딩해보기 (pseudo code)
    - 상태1$*$ : `색상` (e.g. red)
    - 상태2 : `종류` (e.g. sports car)
    - 동작1 : `주행`
    - 동작2 : `가속` 
    
    
    
$*$ 교재에는 '상태'라는 말 대신 '속성'이라고 되어 있지만, 위의 표의 대응관계에 따르면 '속성'이라는 말이 더 적절할 것 같아서 수정했다.

```python
class Car:
    color = 'red'
    category = 'sports car'

    def drive(self):
        print("I'm driving")

    def accel(self, speed_up, current_speed=10):
        self.speed_up = speed_up
        self.current_speed = current_speed + speed_up
        print("speed up", self.speed_up, "driving at", self.current_speed)
a = Car()
print(a.color)
print(a.category)
print(a.drive())
print(a.accel(5))
print(a.accel(5,10))
print(a.accel(5,20))
```

<font color = red> `__init__`과 같은 것이 없는데도 왜 잘 동작하지?</font>

위 코드에 대한, 주석이 달린 버전



지금까지 나에게 있어 `class`란 '어려운 어떤 것'이었다.
사실 지금도 어렵다.

하지만 위의 코드는 완전하게 이해했다. (고 생각한다.)
단 한가지, 빨간 색으로 위에 적었지만 `_init__`없이도 저것들이 동작하는 이유를 잘 모르겠다. 뒤에 나오니 그때가서 이해하면 되겠다.
<font color=green> 이제는 이해했다.</font>

거의 완전하게 이해했지만, 또 한가지 어려울 수 있는 부분은 `self`이다.
코드를 다시 보자.

```python
class Car:
    color = 'red'
    category = 'sports car'

    def drive(self):
        print("I'm driving")

    def accel(self, speed_up, current_speed=10):
        self.speed_up = speed_up
        self.current_speed = current_speed + speed_up
        print("speed up", self.speed_up, "driving at", self.current_speed)
a = Car()
print(a.accel(5))
```

그리고 정리해서 적어보자.
 - `self`는 *attribute*를 정의할 때는 나오지 않고 *method*를 정의할 때 나온다.
 - *method*는 소괄호를 사용해 인자(argument)를 받게 되는데, 어느 *method*이건 항상 `self`라는 인자를 받게 된다.
 - 이때 *self*는 해당 클래스의 인스턴스인 객체의 의미이다.
 - 즉, `a = Car()`라는 명령어로 `Car`의 인스턴스이자 객체인 `a`를 선언했었다.
 - 여기서 `self`의 의미는 `a`가 되는 것이다.
 - 따라서 `self.speed_up`의 의미는 `a.speed_up`의 의미가 된다.

더 정리할 수 있다.
하지만 더 정리하면 오히려 이해가 되지 않는 느낌이다.
그러니 여기서 그만하겠다.

다음은 교재 본문의 마지막 내용이다.

<blockquote>
    
`self`는 자기 자신입니다.

클래스에 의해 생성된 객체(인스턴스)를 가리킵니다.

클래스의 메서드는 인자로 해당 인스턴스(`self`)를 받아야 합니다.

메서드를 호출할 때는 `self` 인자를 전달하지 않습니다.
`self`의 값은 인터프리터가 제공합니다.

인스턴스 변수를 정의할 때에는 접두사 self.을 붙여줍니다.

</blockquote>


## 7.6. 클래스 기본 문법 (3) 생성자

`__init__` : 인스턴스 객체의 속성값을 초기화한다.

아까 만들었던, 클래스를 정의하는 코드는 문제가 있었다.

```python
class Car:
    color = 'red'
    category = 'sports car'

    def drive(self):
        print("I'm driving")

    def accel(self, speed_up, current_speed=10):
        self.speed_up = speed_up
        self.current_speed = current_speed + self.speed_up
        print("speed up", self.speed_up, "driving at", self.current_speed)
```

아니.. 딱히 문제는 없긴 한데, 개선할 수 있는 사항이 있다.
뭐냐하면, 너무 제한된 요건에서만 `Car` 클래스를 사용할 수 있다는 점이다.
차 색깔은 무조건 빨간색이어야 하고, 차 종류는 무조건 스포츠카여야만 한다.
하지만, 그래도 명색이 `class`이름이 `Car`인데 파란색 차를 호출할 수도 있으면 좋을 것 같고, SUV를 호출할 수도 있으면 좋을 것 같다.

그렇게 하자면 `Car` 자체에 몇 개 인자를 부여하는 방법을 사용할 수 있을 것이다.
다음과 같이 코드를 짜면 될까?
서로 다른 두 개를 시도해보았다.

```python
class Car2(color='red',category='sports car'):
    def drive(self):
        print("I'm driving")

    def accel(self, speed_up, current_speed=10):
        self.speed_up = speed_up
        self.current_speed = current_speed + self.speed_up
        print("speed up", self.speed_up, "driving at", self.current_speed)
```

```python
class Car2(color,category):
    color='red'
    category='sports car'
    def drive(self):
        print("I'm driving")

    def accel(self, speed_up, current_speed=10):
        self.speed_up = speed_up
        self.current_speed = current_speed + self.speed_up
        print("speed up", self.speed_up, "driving at", self.current_speed)
```

에러가 발생한다.
이 문법으로 쓰면 안된다는 것이다.
옳게 된 코드는 다음과 같다.

```python
class Car2:
    def __init__(self, color, category):
        self.color = color
        self.category = category

    def drive(self):
        print("I'm driving")

    def accel(self, speed_up, current_speed=10):
        self.speed_up = speed_up
        self.current_speed = current_speed + self.speed_up
        print("speed up", self.speed_up, "driving at", self.current_speed)
```

그러니까, `Car2` 클래스를 선언하면서 바로 중괄호를 쓰는 것은 문법에 옳지 않다.
(아마 이러한 중괄호는 상속의 의미로 쓰일 것이다. 맞나?<font color = blue> 맞다.</font>)

대신, 아래애 `__init__`이라는 이름의 메소드를 선언하면서, 거기에 중괄호로 color와 category를 지정할 수 있다.

'6.11'의 이름규칙에서 변수명의 앞에 underscore(\_)를 적는 것은 함수 내부 변수를 일컫는 것이고, 뒤에 적는 것은 파이썬 기본 키워드와의 중복을 피하기 위함이라고 했었다. 그러면 underscore를 두 번 적는 것(\_\_)은 일반적으로 어떤 의미일지도 궁금하다. (이때 나오겠지)
<font color = green> --> 이것은 던더(Double Under)라고 발음하고, 매직 매서드라고 한다. 이에 대해서는 아직 더 공부하지 않겠다. </font>

`__init__`은 단순히 문법적인 규칙의 의미만 있는 것은 아니다.
init, 즉 initialize(초기화)시킨다는 것이다.
무엇을 초기화하냐 하면 객체의 *속성값(attribute)*을 초기화한다고 되어 있다.

그러니까 사실, 나는 위의 두 코드를 통해 속성초기화가 되어 있지 않은 class를 만들어보고 부적절하다고 쓰려고 했다.
그런데 일단 동작하지 않으니 적당한 반례 코드를 제시하는 건 지금은 못하겠다.

다시 위의 코드를 살펴보자.
 - 정리하면, `class`에는 *attribute*와 *method*라는 중요한 두 개념이 있는데, 둘다 `def`절로 시작한다.
`def` 다음에는
    - *attribute*의 경우에는 `__init__`을 적고
    - *method*의 경우에는 해당 메소드 이름을 적는다.
 - *attribute*
    - `def __init__` 다음에는 기본 인자로서  `self`를 받고, 추가 인자(arguement)로서 *attribute*들을 받는다.
    - 이때, `self`는 해당되는 객체(*object*)의 의미이다.
    -  그리고, <font color=blue> 나중에 </font> *object.attribute*의 형식으로 사용할 수 있도록  
    `self`.*attribute*=*attribute*
    의 형태로 선언한다.
    - 이 선언에서 앞의 단어와 뒤의 단어가 꼭 같을 필요는 없다. 다른 단어나 다른 형태로 설정하더라도, 적당히 문법만 맞추면 문제가 없다.
 - *method* 도 거의 같지만 중복되는 내용을 포함해 다 적겠다.
    - `def `*method*` ` 다음에는 기본 인자로서  `self`를 받고, 추가 인자(arguement)로서 *arguement*들을 받는다.
    - 이때, `self`는 해당되는 객체(*object*)의 의미이다.
    -  그리고, <font color=blue>함수 내에서</font> *object.attribute*의 형식으로 사용할 수 있도록  
    `self`.*argument*=*arguement*
    의 형태로 선언한다.
    - 이 선언에서 앞의 단어와 뒤의 단어가 꼭 같을 필요는 없다. 다른 단어나 <font color=blue>다른 형태</font>로 설정하더라도, 적당히 문법만 맞추면 문제가 없다.
    - 이후에는 이 메소드의 동작을 적는다.

```python
#인스턴스 객체 선언
car1 = Car()
car21 = Car2('yellow', 'sedan')
car22 = Car2('green', 'suv')
print(car1.color)
print(car21.color)
print(car22.color)
```

위와 같이 인스턴스 객체의 속성값이 매번 초기화되었다.

바로 다음 노드에 나오는 내용이지만, Car()에서 정의하는 `color`와 `category`는 클래스 변수에 해당한다.
또한 Car2에서 정의하는 `color`와 `category`는 인스턴스 변수에 해당한다.
그러니까 `car1`의 정의와 `car21`의 정의 사이에는 초기화가 일어나지 않았다고 봐야 할 것 같다.
대신 `car21`과 `car22` 사이에는 초기화가 일어났다.

## 7.7. 클래스 기본 문법 (4) 클래스 변수와 인스턴스 변수

```python
class Car:
    Manufacture = "India"

    def __init__(self, color, category='sedan'):
        self.color = color
        self.category = category
```

에서 `Manufacture`는 클래스변수이고 `color`, `category`는 인스턴스 변수이다.

```python
car1 = Car('red','sports car')
car2 = Car('white')
print(car1.Manufacture, car1.color, car1.category)
print(car2.Manufacture, car2.color, car2.category)
```

에서
 - 클래스 변수(`Manufacture`)는 인스턴스(`car1`인지 `car2`)에 depend하지 않고 독립적인 값으로서 존재한다.
 다만, 당연히 class의 종류에는 depend한다.
 - 인스턴스변수(`color`, `category`)는 당연히 class 종류에도 당연히 depend하지만, 인스턴스(`car1`인지 `car2`)에도 depend한다.
