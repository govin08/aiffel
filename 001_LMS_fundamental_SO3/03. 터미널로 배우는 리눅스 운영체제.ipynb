### 3.8. 운영체제 일반 개론 (4) 프로세스와 스레드
 - **프로그램** : 컴퓨터가 수행할 작업을 기록해놓은 것
 - 프로그램이 실제로 작업으로서 수행될 때, 운영체제는 이를 **프로세스(Process)**라는 단위로 관리한다.
 - 즉 운영체제는 프로세스 단위로 메모리와 CPU 연산 등 컴퓨팅 자원을 분배한다.
  - **메모리 관리**와 프로세스 격리
    - 운영체제는 프로세스가 컴퓨터의 가상 메모리로서 일부만 떼어서 제공한다.
    - 다른 프로세스의 메모리를 훔쳐볼 수 없도록 하기 위함이다.
    - 프로세스들이 서로 소통하기 위해서는 프로세스간 통신(IPC, Inter-Process Communication)이 필요하다.
  - **CPU 자원 관리**
    - 메모리 : 주소로서 프로세스에게 제공
    - CPU연산 : 코어 단위로 프로세스에게 제공
    - **스레드**(thread)란, 프로세스가 코어를 사용할 때의 그 단위.
    - 기본적으로, 프로세스는 하나의 스레드를 가지고 있음.
    - 개발자는 프로그램이 더 많은 스레드를 사용하도록 할 수 있음. (다중 CPU 코어)
  - 프로세스 : 운영체제가 메모리를 할당하는 **작업단위**
  - 스레드 : 프로세스가 할당받은 메모리를 활용하는 **실행단위**
  - 멀티프로세싱 : 하나의 프로그램 안에서 여러 개의 프로세스를 활용한다.
  - 멀티스레딩 : 하나의 프로세스 안에서 여러 개의 스레드를 활용한다.
 - 내가 이해한 마인드맵 (부정확할 가능성이 농후하다.)
 ![프로세스와 스레드](https://i.imgur.com/FAVN6qj.jpeg)

### 3.9. 리눅스 기반의 운영체제 (1) 터미널

 - 지금부터 리눅스 기반 운영체제에 특화된 내용들에 집중할 예정이다.
 - `Cloud Shell`은 터미널이라기보다는 **터미널 에뮬레이터**이다.
 - 터미널(terminal)이란
  - 초창기 컴퓨터 시절에 쓰이던 용어.
  - 입출력만 담당하는 CRT 모니터와 키보드로 구성된 말단(terminal)의 기기.
  - 여러 대의 터미널을 한 대에 메인 프레임에 연결할 수 있었음.
  - teletype
 - teletype의 흔적 (tty)
 - 우분투에서의 터미널
  - 우분투에서는 사용자의 그래픽 환경을 2번 터미널에 붙여둔다.
  - 우리는 CLI셸(tty3)에서 무사히 GUI셸(tty2)로 귀환할 수 있다.
  - *즉, 터미널이란 컴퓨터에 정보를 입력하고 출력하는 소프트웨어 및 하드웨어로서의 장치이다.*
  - 우리가 이용하는 터미널 : tty3

### 3.10. 리눅스 기반의 운영체제 (2) 프로세스 관리와 시스템 콜
- 현재 실행중인 프로세스들을 확인한다.

`# ps`

`# ps -e`

`# ps -ef`

![ps -ef](https://i.imgur.com/81t4Cl2.png)
 - 위 캡쳐에 대한 설명
  - PID : process ID
  - TTY : 해당 프로세스가 붙어있는 teletype(터미널)
  - 총 9개의 `tty`가 조회된다.
  - 이중 가상 터미널들은 pts/0, 1, 2, 3, 4의 다섯 개 (혹은 여섯 개)인 것 같다. (pseudo teletype slave)
 - 여러 프로세스들 중 `Xorg`를 죽인다.
 
`# sudo kill -s KILL [PPID]`
 - 직접 해본 실습 및 캡쳐
![sudo kill](https://i.imgur.com/SbI8nx2.png)

`# ps -ef`

`# ps -ef | grep Xorg`

`# sudo kill -s KILL 63`
  - grep : 특정 문자열을 포함한 라인만을 선택하여 출력한다.
  - 파이프(|) : 다른 명령문과 결합해 사용한다.
  - sudo : 최고 관리자 권한으로 실행한다.
  - kill : 죽인다? / PID에 해당하는 프로세스에 시그널을 보낸다.
 - Xorg를 죽인다는 것은, 현재 `Cloud Shell`을 실행하고 있는 프로세스를 죽인다는 말인 것 같다.
 이것을 죽이면 화면이 잠시 꺼졌다가 `Cloud Shell`이 자동으로 재시작된다.
 - 시스템 콜 (System call)
  - 프로세스가 커널에게 요청을 전달하는 기능이다.
  - 운영체제의 커널이 프로그램에게 제공하는 인터페이스이다.
 - `kill`은 프로세스를 죽이는 명령이 아니라, 프로세스에 신호를 보내는 명령어이다.
  - `Ctrl+C` : interupt
  - `Term` : terminate

### 3.11. 리눅스 기반의 운영체제 (3) 프로그램과 환경변수
- 앞서 내용(3.2.)에서 Unix의 철학은 "모든 것은 파일이다"라고 했었다.
 - 따라서 우리가 지금 사용하는 명령어들은 파일로서 저장되어 있다.
 - 예컨대 명령어 `ls` 파일 위치를 조회할 수 있다.
 - `ls`는 실제 프로그램의 일종이기 때문에 이것이 가능하다.
 
`# which ls`

 - 반면, `cd`는 셸에서 제공하는 셸 명령어라서 아무 것도 출력되지 않는다.

`# which cd`

 - `bash`는 우리가 사용하는 셸로서, 하나의 프로그램이다.
 
`# which bash`
 
 (이하 내용들은 잘 이해되지 않는다.)

 - **환경변수(environment variable)**
  - 셸에서 단어 앞에 `$`가 붙으면 환경변수를 의미한다.
  - 환경변수란, 운영체제가 프로세스 단위로 사용하는 변수이다.
  - `PATH`라는 이름의 환경변수(`$`)의 값을 출력(`echo`)한다.
 - 실습
  - `PATH` 환경 변수의 기본값은 사용자가 로그인할 때 운영체제가 자동적으로 정의한다.
  - 하지만, 사용자가 직접 수정할 수도 있다.
  - 셸 명령어인 `export`를 통해 `PATH` 앞에 `/aiffel` 디렉토리를 추가할 수 있다.
  
`# echo $PATH`

`# export PATH=\aiffel:$PATH`

`# echo $PATH`
 - 현재 설정된 환경변수들을 모두 출력한다.

`# env`
 

### 3.12. 리눅스 기반의 운영체제 (4) 패키지 관리자

 - 우분투 운영체제에서 제공하는 APT(Advanced packaging tool)
  - **패키지 관리자(package manager)**의 일종
 - 패키지 관리자
  - 공인 저장소에서 프로그램과 라이브러리를 패키지 단위로 설치하고 관리할 수 있도록 하는 프로그램
  - 그러니까, TeX에서의 CTAN 같은 건가 보다.
 - 다른 패키지 관리자 프로그램들
  - Ubuntu의 Snap
  - CentOS의 YUM (Yellowdog Updater Modified)
  - Red Hat 계열의 RPM (Red Hat Package Manager)
    
    \*파이썬의 PIP (Python Package Manager)
 - APT로 `htop`이라는 프로그램 설치
  - `htop` : CPU 및 메모리 사용량을 시각화하고, 프로세스 별 CPU 사용량을 표시해주는 프로그램
  - Windows의 작업관리자 같은 개념인가보다.
 - 패키지 목록을 최신 버전으로 가져온다.

`# sudo apt-get update`
 
 -  `htop`을 설치한다.
 
`# sudo apt-get install htop`
 
`# htop`

 - `htop`을 삭제한다.
 
`# sudo apt-get remove htop`

#### `sudo` 명령어
 - 사용자와 권한의 개념과 그 중요성
  - 옛날에는 메인프레임 컴퓨터 하나에 여러 개의 터미널을 붙여 사용했다.
  - 사용자(user)를 구분하고 권한을 나누는 일은 중요했다.
  - 특히 UNIX 계열은 서버용 내지는 기타 산업용으로 쓰이는 경우가 잦기에, 사용자와 권한의 개념은 중요하다.
 - 유닉스 계열 운영체제에서 모든 권한은 사용자를 기준으로 인증한다.
 - `ps -ef` : 프로세스마다 누구의 권한으로 실행되는지(`UID`, UserID) 확인할 수 있다.
 - 리눅스 기반 운영체제에서는 각각의 사용자 계정에게 운영체제에 대한 모든 권한이 주어지지는 않는다.
  - 완전한 권한은 **최고관리자(superuser)**에게만 부여된다.
  - 이를 Windows에서는 Administrator라고 부르고,유닉스 계열에서는 `root`라고 부른다.
 - 최고관리자의 권한 빌리기
  - Windows의 "관리자 권한으로 실행"
  - 유닉스 계열의 `sudo`(superuser do)

#### 사용자 그룹
 - 사용자가 여러 명 있을 때, 쉽게 묶어서 관리하는 개념
 - 유닉스계열 운영체제의 철학, "설정파일이나 장치 모두 파일로 접근해야 한다."

  - 컴퓨터에 어떤 그룹들이 있고, 해당 그룹마다 어떤 사용자들이 들어있는지 확인한다. (아래 명령어)

### 3.14. 기타 (1) 약간의 하드웨어 1 : 연산장치
 - 중앙처리장치 : CPU(Central Processing Unit)
  - 냉각팬
  - 실리콘 웨이퍼(반도체)
  - 집적회로 (IC, Integrated Circuit)
  - 포토리소그래피 공정 (photolithography)

#### 기계어
 - **기계어**는 CPU칩이 처리할 수 있는 명령어
 - 우리가 사용하는 프로그래밍 언어(C, C++, Python)와는 다름.
 - 인간이 이해할 수 있는 언어로 짠 프로그램을 CPU에서 실행할 수 있는 기계어로 번역하고 최적화하는 과정이 **컴파일**이다.
 
#### 명령어집합
 - 명령어집합 : Instruction set
 - 회로의 설계는 제조사마다(Intel, AMD 등) 다르지만, 명령어집합은 상당 부분 호환된다.
 - 다만 모바일 기기는 ARM 社의  RISC(Reduced Instruction Set Computer)라는 구조를 가진다.

#### 메모리와의 관계
 - CPU 성능 중 하나인 (32bit / 64bit)는 CPU가 한 번에 읽고 쓸 수 있는 데이터의 크기를 나타낸다.
 - 이것은 사용가능한 **메모리 주소**의 범위에 영햐을 미친다.
 - 예를 들어, 32bit에서는 $2^{32}\approx 42$억 개의 메모리 주소공간만을 사용할 수 있다.
 
#### CPU 성능과 발전
 - CPU 성능 중 또하나로서 클럭이 있다.
 - 클럭(clock)은 초당 처리속도를 의미한다.
 - 하지만 현재 컴퓨터 성능의 발전 추세는, 클럭수를 늘리기보다는, 코어수를 늘리는 쪽으로 발전되는 경향이 있다.

#### 다른 연산장치
 - GPU : Grpahic Processing Unit
  - 코어 수를 획기적으로 늘려 다차원 행렬 연산을 빠르게 처리한다.
  - 8코어 CPU : 박사학위를 소지한 수학자 8명
  - GPU : 사칙연산밖에 못하지만 계산을 틀리지는 않는 초등학생 3,000명
 - TPU : Tensor Processing Unit
  - 구글에서 딥러닝 전용으로 만든 연산장치
 - 기타
  - ASIC : Application-Specific Integrated Circuit, 어플리케이션 특화 집적회로
  - FPGA : Field-Programmable Gate Array

### 3.15. 기타 (2) 약간의 하드웨어 2 : 저장장치
#### 저장장치
 - 저장장치 비교
 |주 저장장치|보조 저장장치|
 |:-:|:-:|
 |RAM|HDD, SSD|
 |비싸다|싸다|
 |전원이 꺼지면 데이터가 사라진다.|전원이 꺼져도 데이터는 보존된다.|
 |처리속도가 상대적으로 빨라서 CPU와 대화(!)가 가능하다.|처리속도가 상대적으로 느려서 말그대로 저장의 목적으로만 쓰인다.|
 
  - RAM : Random Access Memory
  - HDD : Hard Disc Drive
  - SSD : Solid State Drive
 
#### 딥러닝과 저장장치
 - CPU 대신 GPU를 쓰는 건 학습속도를 올릴 수 있다.
 - 즉, 연산장치는 딥러닝에서 중요한 역할을 한다.
 - 하지만 연산장치 말고도 저장장치 또한 중요할 수 있다.
  - 보조저장장치를 HDD에서 SSD로 변경하는 것이 효과적일 수 있다.
  - 주저장장치인 RAM이 적절히 멀티스레딩할 수 있도로 조절하는 것이 효과적일 수 있다.

### 3.16. 기타 (3) 클라우드 가상화

#### 클라우드 컴퓨팅
 - 클라우드 컴퓨팅의 기원 : 아마존
  - 아마존(Amazon)에서 주요 연휴마다 고객들이 몰려 서버가 다운되는 사태가 발생
  - 수요에 따라 서버를 빌렸다가 반납하는 자회사를 설립
  - **AWS (Amazon Web Service)**
 - AWS는 실제 물리적인 서버를 임대하는 것이 아니라, 한 대의 물리적인 서버를 여러 개의 가상 서버로 쪼개어 판매하는 것.

#### 가상화
 - **가상화(Virtualization)**
  - AWS의 예에서처럼 하나의 컴퓨터를 쪼개어 여러 대의 컴퓨터가 있는 것처럼 사용하게 하는 기술
  - **호스트(host)** : 실제 물리적인 서버
  - **게스트(guest)** : 그 안에 돌아가는 가상 서버
 - 가상화의 종류
  - **가상머신(Virtual Machine) 가상화**
  - **컨테이너(Container) 가상화**
 - **도커(Docker)** : 컨테이너 가상화를 손쉽게 제공하는 프로그램
  - (1) 한 곳에서 실행 환경을 컨테이너로 구성해놓는다.
    - 우분투 버전
    - 라이브러리
    - 필요한 프로그램들
  - (2) 클라우드 / 로컬 환경으로 가져와 독립된 작업환경을 구축한다.
 - Anaconda의 virtual environment 도 가상화의 한 종류로 볼 수 있다.
