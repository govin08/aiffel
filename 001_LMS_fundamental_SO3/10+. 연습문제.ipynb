```python
def not_outlier(df, col, z):
    return df[abs(df[col] - np.mean(df[col]))/np.std(df[col]) <= z].index
print("üëΩ It's okay, no biggie.")
```

```python
def outlier_list(df,col_list,z):
    return np.logical_or.reduce([pd.Series(abs(df[col] - np.mean(df[col]))/np.std(df[col]) <= z) for col in col_list])
```

```python
col_list = [ 'NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']
```

```python
outlier_list(games,col_list,1.5)
```
