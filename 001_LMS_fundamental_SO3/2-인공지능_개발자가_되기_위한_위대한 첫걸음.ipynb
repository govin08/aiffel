### 2.5. 가상환경으로 패키지를 조금 더 유연하게 관리해보자.

 - 가상환경
 
 [파이썬 코딩 도장](https://dojang.io/mod/page/view.php?id=2470)
 
 __Q17. 가상환경이란 무엇인가요?__
 
 __Q18. 가상환경은 왜 필요한가요?__
 
 Virtual Environment

 파이썬으로 작업할 때 여러 패키지들을 다운받고, 버전을 맞춰서 작업하게 된다.
 그런데 서로 다른 두 개 이상의 작업을 하게 된다면, 한 개의 컴퓨터 환경에서 그것들을 다 처리하기가 힘들다.
 가령, 프로젝트 A와 프로젝트 B를 수행하는데 A에서는 tensorflow 1 버전으로 작업해놓았고, B에서는 tensorflow 2 버전으로 작업하려 한다면, 동시에 작업하는 것이 불가능하다.

 그럴 때에 가상환경 a와 가상환경 b를 만들어놓아서, 가상환경 a에는 tensorflow 1을, 가상환경 b에는 tensorflow 2를 설치할 수 있다.
 그렇게 하면 가상환경 a에서는 프로젝트 A를 수행할 수 있고 가상환경 b에서는 프로젝트 B를 수행할 수 있게 된다.
 
 - 가상환경을 사용함으로써 패키지 간의 충돌을 예방할 수 있다.
 - 용어 정리
  - Package : 특정 기능을 위한 여러 함수 또는 클래스를 담고 있는 보따리로서, library, module과 비슷한 개념
   - library는 package보다 조금 더 좁은 의미의 개념이라고 보는 것이 옳다.
   - module은 조금 작은 개념이라는 의미로 쓰이기도 한다.
  - virtual environment : 컴퓨터에 설치된 패키지 간의 충돌 또는 패키지 버전에 의한 이슈 등을 방지하기 위해 가상으로 나누어서 사용하는 환경 즉, 특정 프로그램을 돌리기 위해 필요한 패키지들을 모아 만든 각각의 독립된 방과 같은 개념
 - 가상환경을 관리하는 툴 : pyenv / pip / **anaconda**

#### (1) 아나콘다 설치

 [Ubuntu 20.04 - Anaconda 설치 및 가상환경 구축](https://mishuni.tistory.com/118)

 - `conda`의 위치 확인

 - `conda`의 버전 확인

#### (2) 가상환경 생성 및 실행

 [Create virtual environments for python with conda](https://uoa-eresearch.github.io/eresearch-cookbook/recipe/2014/11/20/conda/)

 - `conda`에 `my_env_name`이라는 이름의 가상환경을 생성하되, 파이썬 버전은 3.9.7

 - `conda`에 설치된 가상환경 나열

\# conda env list

 - `conda`를 초기화

\# conda init

 - `my_env_name`을 활성화

\# conda activate my_env_list

 - `tensorflow`가 설치되어 있는지 조회

\# pip list | grep tensorflow

\# pip install tensorflow==2.6.0

 - 비활성화

\# conda deactivate

 - 가상환경 `my_env_name`을 제거

\# conda env remove -n my_env_name
