## 7.2. 객체 지향 프로그래밍 (1) 객체란?

### 객체가 등장하기까지

 - 컴퓨터의 초기 시대 (1950년대)
    - 점점 더 복잡한 일을 처리해나가기 시작했다.
 -  변수, 함수 정도가 쓰였다.
 - '컴퓨터에게 해야 할 일을 정확히 적어줘야 했다.
    - 조금만 복잡한 일을 하려고 해도 프로그램이 상당히 복잡해져야 했다.
 - 변수들을 효율적으로 다루고 잘 엮고 구조화하기 위한 여러가지 개념들이 등장했다.
    - 구조체(structure)
    - 객체(object)
    - 클래스(class)
 - 또다른 이슈 : 변수와 함수를 어떻게 엮을 것인가?
    - 메시지(message)
    - 메서드(method)
    - 멤버 함수(member function)
 - [LISP 프로그래밍 언어](https://web.archive.org/web/20100717111134/http://history.siam.org/sup/Fox_1960_LISP.pdf)
  - 이후 이 아이디어들은 '객체'라고 하는 이름의 패러다임으로 통용되었고 자주 사용되었다.
  
  
  
  - 도식에서, 내가 이미 들어보거나 알고 있는 개념은 보라색으로 표시하였다.
  - **객체의 절대적 정의가 없다.**
    - 객체라는 개념은 계속해서 바뀌고 진화해간다.
    - 같은 개념을 다른 말로 쓰기도 하고, 다른 개념을 같은 말로 쓰기도 한다.
    - 프로그래밍 언어마다도 개념이나 용어가 다를 수 있다. 

### 객체
> Everything in Python is an object, and almost everything has attributes and methods.
 
 - 간단히 말하면, 클래스 안에서
    - 변수를 만들면 그것은 속성(attribute)이고, 
    - 함수를 만들면 그것은 메서드(method)가 된다.
    - '컴퓨터의 세상은 생각보다 엄밀하게 정의하지 않아요.'
 - `bool`, `int`, `float`, `str`, `list`, `dict`, `function`, 모듈, 프로그램 등이 모두 객체이다.
     - (모듈과 프로그램은 직접 본 적이 없어서 따로 표시하지 않았다.)
 - 다른 언어에서는 모든 것이 객체가 아닐 수도 있다.
     - 객체가 없는 언어
     - 일부분만 객체인 언어
     - 파이썬의, '모든 것은 객체'라는 것은 하나의 언어로서의 특징이지, 장점이라고 하기는 어려울 수 있다.

### 변수

 - `myvar = 3`
     - `=` 연산자를 이용해 값을 할당하는 것
     - 복사하는 것이 아니라, 데이터가 담긴 객체에 **그냥 이름을 붙이는 것**이다.
     - **참조**할 수 있게 하는 것
 - `myvar` : 3을 담은 객체
     - `3` : 의미적으로는 `3이라는 고유한 값.`이라고 이야기할 수 있겠지만, 정확하게는 `3` 또한 객체이다.
     - 즉 `3`은 정수 타입의 객체이다.
 - `myword = 'cat'`
     - `cat`이라는 문자열 타입의 객체를 생성하고, `myword`라는 변수(객체)에 할당하여 문자열 객체를 참조하게 한다.
 - `myword.upper()` / `cat.upper()`
     - 객체 `myword`/`cat`의 매서드
 - `id()`
     - 파이썬의 내장함수(`bulletin_function`)
     - 고윳값(identity)을 반환한다. 고윳값이란
         - 객체마다 부여된 유일한 값 혹은
         - 파이썬에서는 '메모리주소'라고 이해하면 된다.
     - 변수를 할당하고 (`var = 4`) `var`과 `4`의 고윳값을 각각 계산하면 같은 결과가 나온다.
     이것은 `is` 명령어를 통해서도 확인할 수 있을 것이다.

```python
var = 4
print(type(id))
print(id(4))
print(id(var))
```

### 얕은 복사, 깊은 복사

```python
mylist = [1, 2, 3]
var = mylist
var
```

```python
import copy
A = [[1,2],[3,4]]
B = A # 할당 (assign)
C = A.copy() # 얕은 복사 (copy)
D = copy.deepcopy(A) #깊은 복사 (deep copy)

print('A :',A)
print('B :',B)
print('C :',C)
print('D :',D)
print()
print('B == A :',B == A)
print('C == A :',C == A)
print('D == A :',D == A)
print()
print('id(A) :',id(A))
print('id(B) :',id(A))
print('id(C) :',id(A))
print('id(D) :',id(A))
print()
print('B is A :',B is A)
print('C is A :',B is A)
print('D is A :',B is A)
print()
```
