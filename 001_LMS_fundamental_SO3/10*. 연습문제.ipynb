원래 `trade.csv`에서는 `how='any'`로 썼었다.
그때는, 수출건수, 수출금액, 수입건수, 수입금액, 무역수지 열에 결측치가 있었다.
먼저 `dropna(how='all')`로 모든 열이 결측치인 열을 제거했다.
그리고 나니 결측치가 있는 행은 하나의 행밖에 남지 않았고, 해당 행에는 하나의 열에 해당하는 값만 `NaN`이었다.
그리고 그 결측치를 적절히 보간하여 사용했었다.

이번 `vgsales.csv`에서는 `how='any'`로 하여 결측치가 있는 열을 모두 날려버렸다.
이번에는 `Year`와 `Publisher`에 결측치가 있었는데, 두 열이 모두 결측치인 열을 제거하더라도 상당히 많은 행이 처리되지 않은 채로 남았다.
아까 무역수지처럼, 연속적인 숫자를 갖는 값이라면 어떻게든 보간을 하겠는데, 이번에는 게임의 출시년도인 `Year`와 출시 회사인 `Publisher`이다.
이것은 달리 보간할 방법도 없고, 보간한다고 해도 별 의미가 없을 것 같아서, 모두 지웠다.
그런데 모두 지워도 될 것 같은게, 아까에 비해 데이터수가 꽤 많으므로, 괜찮을 것 같다.

## 10.4. 이상치(Outlier)

```python
games
```

```python
def outlier(df, col, z):
    return df[abs(df[col] - np.mean(df[col]))/np.std(df[col])>z].index
print("👽 It's okay, no biggie.")
```

```python
def not_outlier(df, col, z):
    return df[abs(df[col] - np.mean(df[col]))/np.std(df[col]) <= z].index
print("👽 It's okay, no biggie.")
```

```python
def outlier_list(df,col_list,z):
    A = [outlier(df,col,z) for col in col_list]
    np.logical_or.reduce(A)
    return outlier(df,)
```
