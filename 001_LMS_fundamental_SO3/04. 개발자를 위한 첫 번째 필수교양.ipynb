## 4.5. 내 코드의 모든 발자취를 기록할 수 있을까? (2) Git과 GitHub

 - **Git** : 개발을 진행하며 작성하는 소스코드가 업데이트 되는 버전을 기록해두고 관리할 수 있는 소스코드 버전 관리 시스템
 - **GitHub** : Git으로 관리하는 프로젝트를 호스팅하고, 시간과 공간의 제약 없이 협업할 수 있는 온라인 서비스
 - Git이 버전기록을 저장한다면, GitHub에서는 그 기록을 다른 사람과 함께 공유하며 협업할 수 있다.
 - 로컬에서 작업한 내용을 Git에 저장해 두었다면, 그 기록을 온라인 작업공간인 GitHub에 올려 워너격으로도 작업할 수 있도록 한다.

## 4.7. GitHub에 첫번째 잔디 심기! 어렵지 않아요! (1) 로컬 저장소

### 가. 로컬의 Git에 GitHub의 계정 정보 등록하기

`# git config --global user.email "govindah08@gmail.com`

`# git config --global user.name "govin08`

`# git config -l`
 
 - 여기에서 한 것은 제목에서 보듯 '로컬의 Git에 GitHub의 계정정보를 등록'한 것이다.
 즉 Git과 GitHub를 연결한 것이다.
  - 하지만, '로컬의 Git'이라는 것의 실체가 무엇인지는 잘 모르겠다.
  - 교재에는 'Git으로 무얼 할 수 있는지'에 대해서만 설명하고 있고 'Git이 무엇인지', 로컬에 저장되어 있다면 그 'Git을 어떻게 조회하거나 열람할 수 있는지'는 설명되어 있지 않다.
 - 아직 로컬 저장소를 만들지 않았다.
  - 그런데도 'Git과 GitHub를 연결'한다는 것을 보니 이미 '로컬의 Git'은 존재한다고 봐야 할 것 같다.
  내 컴퓨터 어딘가에 Git이라고 하는 게 있다고 생각하면 될 것 같고, 로컬 저장소를 만들기 전임에도 불구하고 Git과 GitHub를 연결할 수 있다고 이해하고 있다.

 - 이 `workspace` 폴더는 로컬 저장소의 역할을 할 것 같다.


### 다. Git으로 버전 관리 시작하기

`# git init`

`# ls -a`

`# cd .git`

`# ls `



 - 폴더 `workspace`로 이동해 Git을 초기화했다.

## 4.8. GitHub에 첫번째 잔디 심기! 어렵지 않아요! (2) 변화 추적하기
### 라. `README.md` 파일 생성하기

`# cd ~/aiffel/workspace`

`# echo "# first-repository"` >> README.md

`# ls`

`# cat README.md`

 - `workspace` 폴더에 가서 `README.md` 파일을 만들었다.
     - 앞으로 이 파일을 GitHub에 올릴 것이다.
 - 그 과정에서 리눅스 명령어인 `cat`을 사용했다.
     - 저렇게 `>>`를 쓰면 왼쪽의 텍스트를 오른쪽의 파일에 쓰겠다는 뜻이다.
     (내가 기억하기로는) 만약 `>`를 쓰고, 왼쪽에 텍스트를 쓰지 않은 채로 오른쪽에 파일만 쓰고 엔터를 누르면 해당 md파일 혹은 txt 파일에 입력할 텍스트를 입력하는 모드가 나타난다.

### 마. git으로 변화 확인! 지금 버전에 도장 쾅! 찍기
`# git status`

`# git add README.md`

`# git commit -m "new readme file"`

- `# git status`는 현재 Git의 상태를 출력한다.

     - 어떤 branch에 속해있는지
     
     - commit할 것이 남아있는지 (이 경우 `Unracked files:`로 시작하는 알림이 뜬다.)
     
     등을 알려준다.

- `# git add README.md`와 `# git commit -m "new readme file"`를 통해, `add`와 `commit`을 실시한다.
    - `add` 시에는 `# git add .`도 유용할 것이다.
    - 또한, 둘을 한꺼번에 하는 명령어도 있다.
- 여하튼 두 명령어를 실행시키고 나면 작업트리(working directory)에 있는 '변경사항'들이 인덱스(index, staging area)를 거쳐 저장소(head repository)에 올라간다.
    - 그러니까, 여기서 말하는 '저장소'는 우리의 `workspace` 폴더와 다르다.
    - 굳이 말하자면, `workspace`폴더의 '변경사항'이 작업트리에 해당한다고 봐야 할 것 같다.
    - 그리고, 별도로 존재하는 '저장소(head repostiory)'라는 개념이 있어서 여기에 그 '변경사항'들이 올라간다.
    - 이때, 저장소(head repositry)는 아직 로컬의 영역이다.

### 바. GitHub에 나의 첫 번째 레파지토리 만들기
 - GitHub 사이트에 들어가 repository를 만든다.

 따라서 'aiffel' repository의 주소는 [https://github.com/govin08/aiffel](https://github.com/govin08/aiffel)이 될 것이다.
 - 현재 instruction 대로 따라오고 있다면 이 repository에는 아무 것도 없어야 한다.

### 사. 내 로컬 저장소와 원격 저장소를 연결해보자!

`# cd ~/aiffel/workplace`

`# git remote add origin https://github.com/govin08/aiffel.git`

 - 지금까지의 상황은
    - (로컬) Git이 구축되어 있고 (`workspace`폴더 등)
    - (원격) GitHub 환경도 마련되어 있다. ('aiffel'이라는 이름의 repository)
 - 이 두 개의 저장소, 즉 로컬 저장소와 원격저장소를 연결해보는 것이 위의 명령어들이다.
 - 이때, `# git remote` 이하의 명령어는 한 번만 실행할 수 있다
    - 이후에 실행하면 다음과 같은 오류 메시지가 뜬다.
    - `fatal: remote origin already exists.`
    - 이 상황이 fatal씩이나 되는 것일까? 경고 메시지가 무섭다. 하지만 별 일 없는 것이라고 정신승리해보자.
 - 두번째 명령어를 입력할 때에, aiffel repository 주소인
     - https://github.com/govin08/aiffel
     - 외에 `.git`이 붙어서
     - https://github.com/govin08/aiffel.git
     - 이 되었다. 이 둘은 서로 다른 것일까?
     - 실제 두 링크를 클릭해보면 같은 웹페이지(https://github.com/govin08/aiffel)로 들어가지는 것 같다.
     - 또한, https://github.com/govin08/aiffel 에서 code 버튼(초록색)을 눌렀을 때 나오는 주소가 바로 https://github.com/govin08/aiffel.git 이다.
     - code 버튼(초록색)을 눌렀을 때 'clone'이라는 단어가 나오는 것으로 봐서는, 아마도 `clone`을 할 때에 두번째 주소를 입력해야하는 것 같다.
     - 한용님 말씀으로는 `git remote`시 반드시 `.git`이 붙은 url을 입력해야 한다고 하셨던 것 같은데, 정확히 모르겠다.

    - 이때 토큰이란, 비밀번호의 역할을 하는 것이다.

    - 하지만, 퍼실님 말씀에 의하면 보안상의 이유로 이렇게 비밀번호 대신 토큰을 사용하는 것으로 바뀌었다고 한다.
 - 토큰이란, 특정한 문자열을 말한다.
    - 깃허브에서는 이 토큰을 자동적으로 생산해준다.
    - 사용자는 이것을 잘 저장해놨다가 (메모장 등) 필요할 때에 사용해도 되고
    - 아니면 다음 subsubsection에서 보는 바와 같이 crendetial.helper를 사용하여도 된다.
    - 하지만 잘 이해가 가지 않으므로, 이에 대해서는 단순히 코드를 나열하는 것으로 갈음하겠다.

### 아. 로컬 저장소의 기록을 원격 저장소로 전송하기

`# git config credential.helper store`

`# git push origin master`
 
 - 주된 명령어는 두번째의 `# git push origin main`이다.
    - 이 명령어를 통해 `push`를 진행한다.
    - 교재에는 `main`이라고 되어 있지만 나는 `master`가 통한다.
    - 그래서 여기에서도 master라고 적어놓았다.
    - Username을 적으라는 란이 뜨면 'govin08'이라고 입력한다.
    - Password를 적으라는 란이 뜨면 토큰을 복사해다가 붙여넣기한다.
    - 붙여넣기를 할 때에는 `Shift+Instert`를 사용했던 것 같은데, 실제로 `Ctrl+V`나 `Ctrl+Shift+V`가 먹히는지는 잘 모르겠다.

`# git config credential.helper "store --file ~/aiffel/.git-credentials"`

`# git config -l`

 - 토큰을 포함한 계정정보를 적당히 저장해놓는 방법을 설명하는 것 같다.
 - `aiffel` 폴더 하위에 `.git-credentials` 파일에 계정정보가 담긴다는 말인 것 같다.
