## 15.3. 확률

- 공리적 정의
  - 표본공간 $\Omega$에 대하여, $\Omega$의 부분집합들의 집합 $\Sigma$가 다음의 세 가지 성질을 만족시키면 $\Sigma$를 $\sigma-$algebra라고 한다.
  $\Sigma$의 원소를 *사건(event)*이라고 부른다.
    - $X\in\Sigma$
    - $\Sigma$가 countable union(countable개의 합집합)에 대하여 닫혀있다.
    - $\Sigma$가 complement(여집합)에 대하여 닫혀있다.  
    
  - 함수 $P:\Sigma\to[0,1]$이 다음 성질을 만족시키면, $P$를 *확률함수*라고 부른다.
    - $P(\Omega)=1$
    - $P$는 countably additive하다.
  - ordered triple $(\Omega,\Sigma,P)$을 확률 공간(probability space)이라고 부른다.
    - 즉, 확률공간이란, measure space $(X,\mathscr M,\mu)$에서 $\mu(X)=1$의 조건만 추가시킨 것이다.

- 상대도수 정의(통계적 정의)
  - 시행이 $n$번 수행되었을 때, 사건 $A$가 $n_A$번 일어난다면, 사건 $A$에 대한 확률을 다음과 같이 정의한다.
  
  $$P(A)=\frac{n_A}n$$

- 고전적 정의
  - 실험없이, 선험적(a priori)으로 정의된다?
  - measure $\mu$에 대하여, $P(A)=\frac{\mu(A)}{\mu(\Omega)}$인 것으로 이해했다.

## 15.8. 전확률과 베이즈 정리

### 분할(partition)
$\{A_1,\cdots,A_n\}$이 pairwise disjoint이고, $A_1\cup\cdots\cup A_n=\Omega$이면 $\{A_1,\cdots,A_n\}$은 $\Omega$의 분할이다.

### 전확률 정리(the law of total probability)  
$\{A_1,\cdots,A_n\}$가 $\Omega$의 분할이면, 사건 $B$에 대하여 다음 식이 성립한다.

\begin{align*}
P(B)
& = P\left(\Omega\cap B\right)\\
& = P\left(\left(\bigcup_{i=1}^nA_i\right)\cap B\right)\\
& = P\left(\bigcup_{i=1}^n(A_i\cap B)\right)\\
& = \sum_{i=1}^n P(A_i\cap B)
\end{align*}

### 베이즈 정리 (Bayes' theorem)  
두 사건 $A$, $B$에 대하여 다음 식이 성립한다.

$$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$$

- $P(A|B)$는 posterior probability (사후확률)라고 불린다.
- $P(A)$는 prior probability (사전확률) 라고 불린다.
- $P(B|A)$는 likelihood function (우도) 이라고 불린다.
<!-- - $P(B)$는 marginal likelihood라고 불리고, $B$는 evidence 라고 불린다. -->
 
$\{A_1,\cdots,A_n\}$가 $\Omega$의 분할이고, $B$가 사건이면, 전확률정리와 베이즈 정리에 의하여 다음 식이 성립한다.

$$P(A_i|B) = \frac{P(B|A_i)P(A_i)}{\sum_{j=1}^nP(A_j\cap B)}$$

베이즈 정리를 사용할 때에는 주로 아래 식을 활용한다.

$$P(A_i|B) \propto P(B|A_i)\times P(A_i)$$

## 15.9. 독립사건, 종속사건, 배반사건

### 독립사건과 종속사건

두 사건 $A$, $B$에 대하여 $P(A) = P(A|B)$이면, $A$와 $B$는 서로 독립이라고 말한다.

아래 조건들은 두 사건 $A$, $B$가 독립이기 위한 필요충분조건들이다.
- $A$, $B$가 독립이다.
- $P(A) = P(A|B)$
- $P(A) = P(A|B) = P(A|B^c)$ : 사건 B가 일어나건 말건 $P(A)$는 일정하다.
- $P(B) = P(B|A)$
- $P(B) = P(B|A) = P(B|A^c)$ : 사건 A가 일어나건 말건 $P(B)$는 일정하다.
- $A$, $B^c$가 독립이다.
- $A^c$, $B$가 독립이다.
- $A^c$, $B^c$가 독립이다.
- $P(A)P(B) =  P(A\cap B)$

두 사건이 $A$, $B$에 대하여 $A$, $B$가 독립이 아니면, $A$, $B$는 서로 종속이라고 말한다.

사건 $A_1,\cdots,A_n$에 대하여, 임의의 $i$, $j$에 대해
$$P(A_i\cap A_j) = P(A_i)P\times(A_j)$$
이면, 사건 $A_1,\cdots,A_n$은 pairwisely independent하다, 라고 말한다.

사건 $A_1,\cdots,A_n$에 대하여, 임의의 $\{i_1,\cdots,i_m\}\subset\{1,\cdots,n\}$에 대해
$$P\left(\bigcap_{k=1}^mA_{i_k}\right) = \prod_{k=1}^mP(A_{i_k})$$
이면, 사건 $A_1,\cdots,A_n$은 mutually independen하다, t라고 말한다.

mutually independent이면 pairwisely independent하다.
하지만 pairwisely independent하다고 해서 mutually independent한 것은 아니다. ([반례](https://en.wikipedia.org/wiki/Independence_(probability_theory)#Pairwise_and_mutual_independence))

### 배반사건

두 사건 $A$, $B$에 대하여 $A$와 $B$가 서로소(disjoint)이면, 즉 $A\cap B=\varnothing$이면, 두 사건 $A$, $B$가 서로 배반사건(exclusive / disjoint)이라고 말한다.

사건 $A_1,\cdots,A_n$에 대하여, 임의의 $i$, $j$에 대해

$$A_i\cap A_j=\varnothing$$

이면, 사건 $A_1,\cdots,A_n$은 pairwisely exclusive / mutually exclusive / pairwisely disjoint / mutually disjoint하다, 라고 말한다.

사건 $A_1,\cdots,A_n$에 대하여
$$\bigcap_{i=1}^mA_{n} = \varnothing$$
이면, 사건 $A_1,\cdots,A_n$은 exclusive / disjoint하다, 라고 말한다.

## 15.10. 확률변수

### 확률변수의 정의

함수 $X:\Omega\to\mathbb R$을 *확률변수*라고 부른다.

조금 더 정확하게 말하려면, $X$가 함수이기만 해서는 안되고, measurable function이어야 한다.
즉 임의의 Borel set $B\subset\mathbb R$에 대하여 $X^{-1}(B)\in\Sigma$여야 한다.

### 확률변수 예시

두 개의 동전을 차례대로 던질 때에 확률공간 $(\Omega, \Sigma, P)$는 다음과 같이 정의된다.
- $\Omega = \{(H,H), (H,T), (T,H), (T,T)\}$
- $\Sigma = 2^\Omega$
- $P(A) = \frac{n(A)}{n(\Omega)}$

확률변수 $X$를 다음과 같이 정의하자.

\begin{align*}
X(H,H)&=1\\
X(H,T)&=2\\
X(T,H)&=3\\
X(T,T)&=4
\end{align*}

그러면, 다음과 같은 계산을 해볼 수 있다.

\begin{align*}
&P(X = 1) = \frac14,
&&P(X = 2) = \frac14,
&&P(X = 3) = \frac14,
&&P(X = 4) = \frac14,\\
&P(X\le2) = \frac12,
&&P(X\ge2) = \frac34,
&&P(1\le X\le2) = 1,
&&P(X<0) = 0.
\end{align*}

확률변수 $X$를 다음과 같이 다르게 정의할 수도 있다. (이게 더 흔히 쓰이는 예시이다.)
두 개의 동전을 차례대로 던졌을 때, 앞면이 나온 횟수를 $X$라고 하자.
즉

\begin{align*}
X(H,H)&=2\\
X(H,T)&=1\\
X(T,H)&=1\\
X(T,T)&=0
\end{align*}

이다.
그러면, 다음과 같은 계산을 해볼 수 있다.

\begin{align*}
P(X=0)&={}_2C_0\left(\frac12\right)^0\left(\frac12\right)^2=\frac14\\
P(X=1)&={}_2C_1\left(\frac12\right)^1\left(\frac12\right)^1=\frac12\\
P(X=2)&={}_2C_2\left(\frac12\right)^2\left(\frac12\right)^0=\frac14
\\\end{align*}

### 표기법(notational convention)

이상에서 $P(X=x)$라는 표현을 썼다.
$P$와 $X$가 각각 $P:\Sigma\to[0,1]$, $X:\Omega\to\mathbb R$인 함수임을 상기해보면, $P(X=x)$의 의미는 다음과 같이 해석하는 것이 정확하다.

\begin{align*}
P(X=x)
&=P\left(X(\omega)=x\right)\\
&=P\left(\{\omega\in\Omega:X(w)=x\}\right)\\
&=P\left(X^{-1}(x)\right)
\end{align*}

## 15.11. 누적분포함수

확률변수 $X:\Omega\to\mathbb R$에 대하여 누적분포함수(CDF, cumulative distribution function) $F_X:\mathbb R\to[0,1]$는 다음과 같이 정의된다.

$$F_X(x) = P(X\le x)$$

누적분포함수 $F_X$의 주요한 성질들은 다음과 같다.

\begin{align*}
&(1)&&x_1<x_2\enspace\Longrightarrow\enspace F_X(x_1)\le F_X(x_2)\\
&(2)&&\lim_{x\to-\infty}F_X(x)=0,\enspace\lim_{x\to\infty}F_X(x)=1\\
&(3)&&F_X(x) = \int_{-\infty}^xf(t)\,dt\\
&(4)&&P(a<X\le b) = F_X(b)-F_X(a)\\
&(5)&&{F_X}'(x) = f(x)
\end{align*}

## 15.12. 이산확률변수와 확률질량함수

### 이산확률변수와 확률질량함수

확률변수 $X$의 image, 즉 $X(\Omega)$가 countable set (at most countable set, i.e., fininte set or countably infinite set)이면 $X$를 *이산확률변수*라고 한다.
만약 $X(\Omega)$가 finite set이면

$$X(\Omega)=\{x_1,x_2,x_3,\cdots,x_n\}$$

이고, countably infinite set이면

$$X(\Omega)=\{x_1,x_2,x_3,\cdots\}$$

이다.
(각각의 $i$에 대하여 $\Omega_i=X^{-1}(x_i)$로 정의하면, $\Omega_i$는 $\Omega$의 partition이 된다.
임의의 $\omega\in\Omega$에 대하여 $\omega\in\Omega_i$이면, $X(\omega)=x_i$가 될 것이다.)
이때, 함수 $p_X:x\to P(X=x)$를 $X$의 *확률질량함수*라고 부른다.

다시 말해, 확률질량함수 $p_X$는 $p_X(x)=P(X=x)$와 같이 정의된 함수이다.
아까의 표기법을 상기하면 $p_X(x)=P\left(X^{-1}(x)\right)$와 같이 쓸 수도 있다.
따라서, $p_X$는 잘 정의된 함수이다.

### 확률질량함수의 성질

확률질량함수 $p_X$에 대하여 다음 식들이 성립한다.

\begin{align*}
&(1) &&p_X(x)\ge0\\
&(2) &&\sum_i p_X(x_i)=1\\
&(3) &&F_X(x) = \sum_{x_i\le x}p_X(x_i)
\end{align*}

(1)은 $P$의 정의로부터 당연하다.

(2)는 $X^{-1}(x_i)$들이 서로 pairwisely disjoint하다는 사실과 $P$의 countable additivity로부터 당연하다.

\begin{align*}
\sum_i p_X(x_i)
&=\sum_iP\left(X^{-1}(x_i)\right)\\
&=P\left(\bigcup_iX^{-1}(x_i)\right)\\
&=P\left(\bigcup_i\Omega_i\right)\\
&=P\left(\Omega\right)\\
&=1
\end{align*}

(3)도 마찬가지로 증명할 수 있다.

\begin{align*}
\sum_{x_i\le x} p_X(x_i)
&=\sum_{x_i\le x}P\left(X^{-1}(x_i)\right)\\
&=P\left(\bigcup_{x_i\le x}X^{-1}(x_i)\right)\\
&=P\left(\bigcup_{x_i\le x}\{\omega\in\Omega:X(\omega)=x_i\}\right)\\
&=P\left(\{\omega\in\Omega:X(\omega)\le x\}\right)\\
&=P(X\le x)\\
&=F_X(x)
\end{align*}


### 예시

아까 동전을 두 개 던지는 예시에서 앞면이 나온 횟수를 $X$라고 하면,

$$X(\Omega)=\{0,1,2\}$$

이므로, $X(\Omega)$는 finite set이다.
그러면, 확률질량함수 $P_X:\{0,1,2\}\to[0,1]$는 다음과 같이 정의된다.

\begin{align*}
P_X(0)&=P(X=0)=\frac14\\
P_X(1)&=P(X=1)=\frac12\\
P_X(2)&=P(X=2)=\frac14
\end{align*}

이것은 위의 조건들 중 (1), (2)를 만족시킴을 쉽게 볼 수 있다.
(3)은 바로 보이지는 않지만, 성립함을 알 수 있다.

## 15.13. 연속확률변수 확률밀도함수

### 연속확률변수와 확률밀도함수

연속확률변수의 정확한 정의는 조금 어렵다.
위키피디아의 설명은 다음과 같다.

> Formally, a continuous random variable is a random variable whose cumulative distribution function is continuous everywhere.

또한, 연속확률변수라는 말이 이산확률변수의 반댓말이라고 생각해서도 안된다.
즉, [이산확률 변수도, 연속확률변수도 아닌 확률변수가 존재한다.](https://math.stackexchange.com/questions/3967681/can-random-variables-be-something-else-other-then-discrete-or-continuous)

나는 다만, 여기서 연속확률변수를 다소 직관적으로, 내가 이해하는 만큼만 정의하려고 한다.
하지만, 이 정의는 formal definition의 subset이 될 것이므로 큰 문제는 없다.

$X(\Omega)$가 실수집합 내의 어떤 구간과 같을때, 즉 $X(\Omega)=[a,b]$이거나 $X(\Omega)=[a,b)$이거나, $X(\Omega)=(a,b]$이거나, $X(\Omega)=(a,b)$일 때, $X$를 *연속확률변수*라고 한다. (단, $-\infty\le a<b\le \infty$)

마찬가지로, 확률밀도함수의 정의도 어렵다.
[위키피디아](https://en.wikipedia.org/wiki/Probability_density_function#Continuous_univariate_random_variable)에 따르면 Radon-Nykodym 정리에 의해 확률밀도함수의 존재가 보장된다고 되어 있다.
그리고 내 생각에는, CDF $F_X$의 존재를 먼저 가정한 다음에, $F_X$가 연속인 경우에 한해 확률밀도함수 $f$를 $F_X$의 도함수로 생각하는 것이 쉬운 접근법일 것 같다.

여하튼, 어려우니까 확률밀도함수가 존재한다고 가정하고 이야기해보자.
확률밀도함수는 다음 식을 만족시키는 함수 $f_X:X(\Omega)\to[0,\infty]$이다.

$$P(\alpha\le X\le\beta)=\int_\alpha^\beta f(x)\,dx.\qquad(a\le \alpha<\beta<b)$$

### 확률밀도함수의 성질

확률밀도함수 $f_X$에 대하여 다음 식들이 성립한다.

\begin{align*}
&(1) &&f_X(x)\ge0\\
&(2) &&\int_a^b f_X(x)\,dx=1\\
&(3) &&F_X(x) = \int_{-\infty}^xf(t)\,dt\\
&(4) &&{F_X}'(x) = f(x)
\end{align*}

## 15.14. 기댓값과 분산 & 조건부 기댓값

### 기댓값과 분산

summation은 intergration의 한 예이다.
즉, 일반적인 Lebesgue 적분에서 counting measure를 선택하면 그것이 곧 summation이 된다.
따라서 연속확률변수를 설명하면서 도입한 integration의 형태만 적어도, summation 형태의 이산확률분포도 다 고려한 셈이 된다.

기댓값과 분산의 정의는 다음과 같다.

\begin{align*}
\mathbb E[X]&=\int_a^bxf_X(x)\,dx\\
\mathbb V[X]&=\int_a^b\left(x-\mathbb E[X]\right)^2f_X(x)\,dx
\end{align*}

한편, 분산은 다음과 같이 계산될 수도 있다.

\begin{align*}
\mathbb V[X]
&=\int_a^b\left(x-\mathbb E[X]\right)^2f_X(x)\,dx\\
&=\int_a^b\left(x^2-2x\mathbb E[X]+\mathbb E[X]^2\right)^2f_X(x)\,dx\\
&=\int_a^bx^2f_X(x)\,dx-2\mathbb E[X]\int_a^bxf_X(x)\,dx+\mathbb E[X]^2\int_a^bf_X(x)\,dx\\
&=\mathbb E\left[X^2\right]-2\mathbb E[X]^2+\mathbb E[X]^2\\
&=\mathbb E\left[X^2\right]-\mathbb E[X]^2
\end{align*}


### 조건부 기댓값

$$\mathbb E[X|Y=y]=\int_a^bxf_{X|Y}(x|y)\,dx$$

동전을 하나 던지는 시행을 생각하자.
확률공간은 다음과 같이 구성된다.
- $\Omega=\{H,T\}$
- $\Sigma=\{\varnothing,\{H\},\{T\},\{H,T\}\}$
- $P(A) = \frac{n(A)}{n(\Omega)}$

이때 동전은 fair coin일 필요는 없고, unfair(biased) coin 일 가능성도 존재한다.
앞면이 나올 확률을 $p$라고 하고 동전의 앞면이 나온 횟수를 $X$라고 하자.
그러면,

\begin{align*}
X(T)&=0\\
X(H)&=1
\end{align*}

이다.
따라서, 확률질량함수 $p_X$는 다음과 같이 나타난다.

\begin{align*}
p_X(0)=p\\
p_X(1)=1-p
\end{align*}

이것을

$$p_X(x)=\begin{cases}p&(x=0)\\1-p&(x=1)\end{cases}$$

로 적어도 된다.
아니면 한 줄로

$$p_X(x)=p^{1-x}(1-p)^x\qquad x=0,\:1$$

로 쓸 수도 있을 것이다.

이와 같이, $n(\Omega)=2$인 시행을 Bernoulli trial이라고 한다.
또한, 이때의 확률분포를 Bernoulli distribution이라고 한다.
임의의 Beroulli trial에 대하여, 일반성을 잃지 않고 $\Omega = \{0,1\}$이라고 두어도 될 것이다.

## 15.16. 이항분포 (1)

동전을 $n$개 던지는 시행을 생각하자.
확률공간은 다음과 같이 구성된다.
- $\Omega=\{H,T\}^n$
- $\Sigma=2^{\{H,T\}^n}$
- $P(A) = \frac{n(A)}{n(\Omega)}$

이번에도 앞면이 나올 확률을 $p$라고 하고, 동전의 앞면이 나온 횟수를 $X$라고 하자.
그러면,

\begin{align*}
X(T,T,T,\cdots,T)&=0\\
&\vdots\\
X(H,H,H,\cdots,H)&=n
\end{align*}

등이다.
확률질량함수 $p_X$는

\begin{align*}
p_X(x) = {}_nC_xp^x(1-p)^{n-x}\qquad x=0,\:1,\:\cdots\:,n
\end{align*}

이다.
$\sum_x p_X(x)=1$이 성립하는지 확인차 계산해보면, 이항정리에 의해

$$\sum_xp_X(x)=\sum_{x=1}^n{}_nC_xp^x(1-p)^{n-x}=\left(p+(1-p)\right)^n=1$$

이다.

어떤 시행을 반복할때, 그 시행들이 독립적이면, 그 일련의 시행들을 통칭해서 *독립시행*이라고 한다.
또한, Bernoulli trial을 $n$번 반복하여$\left(\Omega=\{0,1\},\:p_X(0)=p\right)$, 0이 나온 횟수를 $X$라고 할 때, 확률변수 $X$는 $B(n,p)$의 이항분포를 따른다고 말한다.
기호로는
$$X\sim B(n,p)$$
로 적는다.

```python
# 이항 분포 그림
import math
import matplotlib.pyplot as plt
from functools import reduce
import operator as op

n = 10 # 횟수
p = 0.5 # 성공할 확률

# 좌표 설정
plot_x_range = n
plot_y_range = 0.4

# 조합
def ncr(n, r):
    r = min(r, n-r)
    numer = reduce(op.mul, range(n, n-r, -1), 1)
    denom = reduce(op.mul, range(1, r+1), 1)
    return numer // denom 

# 이항 분포 함수
def p_X(x,n,p):
    return (ncr(n, x))*(p**x)*((1-p)**(n-x)) 

# X값
X = [x for x in range(n+1)]

# 그래프 그리기
plt.figure(figsize=(10,10))
ax1 = plt.subplot(2, 1 ,1)
plt.xlim(0, plot_x_range)
plt.ylim(0,plot_y_range)
plt.title("Binomial distribution with several trials")

ax2 = plt.subplot(2, 1 ,2)
plt.xlim(0, plot_x_range)
plt.ylim(0,plot_y_range)
plt.title("Binomial distribution with several probabilities")

for different_n in [4+3*i for i in range(5)]:
    different_X = [x for x in range(different_n+1)]
    b_y = [p_X(x,different_n,p) for x in different_X]
    ax1.plot(different_X, b_y, label='n='+str(different_n))
    
for different_p in [(1+2*i)/10 for i in range(5)]:
    b_y = [p_X(x,n,different_p) for x in X]
    ax2.plot(X, b_y, label='p='+str(different_p))
    
ax1.legend(loc="best")
ax2.legend(loc="best")
plt.show()
```

## 15.17. 이항분포 (2)

확률변수 $X$가 $X\sim B(n,p)$를 따르면, $X$의 평균 $\mathbb E[X]$과 분산 $\mathbb V[X]$은 다음과 같이 계산할 수 있다.

### 예비 계산

\begin{align*}
r\times{}_mC_r
&=r\times\frac{m!}{(m-r)!r!}\\
&=\frac{m!}{(m-r)!(r-1)!}\\
&=m\times\frac{(m-1)!}{((m-1)-(r-1))!(r-1)!}\\
&=m\times{}_{m-1}C_{r-1}
\end{align*}


###  $\mathbb E[X]$ 계산

\begin{align*}
\mathbb E[X]
&= \sum_{x=0}^nx\times p_X(x)\\
&= \sum_{x=1}^nx\times p_X(x)\\
&= \sum_{x=1}^nx\times {}_nC_xp^x(1-p)^{n-x}\\
&= \sum_{x=1}^nn\times {}_{n-1}C_{x-1}p^x(1-p)^{n-x}\\
&= np\sum_{x=1}^n{}_{n-1}C_{x-1}p^{x-1}(1-p)^{(n-1)-(x-1)}\\
&= np\sum_{y=0}^{n-1}{}_{n-1}C_yp^y(1-p)^{(n-1)-y}\\
&= np\times \left(p+(1-p)\right)^{n-1}\\
&= np
\end{align*}

###  $\mathbb V[X]$ 계산

\begin{align*}
\mathbb E[X^2]
&= \sum_{x=0}^nx^2\times p_X(x)\\
&= \sum_{x=1}^nx\times n\times {}_{n-1}C_{x-1}p^x(1-p)^{n-x}\\
&= np\sum_{x=1}^nx\times {}_{n-1}C_{x-1}p^{x-1}(1-p)^{n-x}\\
&= np\sum_{y=0}^{n-1}(y+1)\times{}_{n-1}C_yp^y(1-p)^{n-1-y}\\
&= np\sum_{y=0}^{n-1}y\times{}_{n-1}C_yp^y(1-p)^{n-1-y}+np\sum_{y=0}^{n-1}\times{}_{n-1}C_yp^y(1-p)^{n-1-y}\\
&= np\sum_{y=1}^{n-1}(n-1)\times{}_{n-2}C_{y-1}p^y(1-p)^{n-1-y}+np\\
&= n(n-1)p^2\sum_{y=1}^{n-1}\times{}_{n-2}C_{y-1}p^{y-1}(1-p)^{n-1-y}+np\\
&= n(n-1)p^2\sum_{z=0}^{n-2}\times{}_{n-2}C_zp^z(1-p)^{n-2-z}+np\\
&= n(n-1)p^2+np\\
\mathbb V[X]
&= \mathbb E[X^2]-\mathbb E[X]^2\\
&= n(n-1)p^2+np - (np)^2\\
&= np-np^2\\
&= np(1-p)
\end{align*}

**Q3.어떤 볼링 선수가 스트라이크를 할 확률이 80%라고 할 때, 이 선수가 3번을 던져 2번 이상 스트라이크를 할 확률을 구해보세요.**

이 볼링 선수가 받은 스트라이크의 개수를 $X$라고 하면, 확률변수 $X$는 이항분포 $B\left(3,\frac45\right)$를 따릅니다.
이 선수가 2번 이상 스트라이크를 받을 확률 $P(X\ge2)$이고, 다음과 같이 계산됩니다.
\begin{align*}
P(X\ge2)
&=P(X=2)+P(X=3)\\
&={}_3C_2\left(\frac45\right)^2\left(\frac15\right)^1
+{}_3C_3\left(\frac45\right)^3\left(\frac15\right)^0\\
&=\frac{48+64}{125}=\frac{112}{125}
\end{align*}

## 15.18. 정규 분포

```python
# 정규 분포 그림
import math
import matplotlib.pyplot as plt

u = 0 # 평균
v = 1# 분산

# 좌표 설정
n = 1000
plot_x_range = 3
plot_y_range = 0.5

# 정규 분포 함수 (직관적 이해를 위해 수식을 그대로 표현)
def f_X(x,u,v):
    return 1/(math.sqrt(2*math.pi*v))*math.exp(-(x-u)**2/(2*v))

# X값
X = [-plot_x_range+2*plot_x_range*x/n for x in range(n+1)]

# 그래프 그리기
plt.figure(figsize=(10,10))
ax1 = plt.subplot(2, 1 ,1)
plt.xlim(-plot_x_range, plot_x_range)
plt.ylim(0,plot_y_range)
plt.title("Normal distribution with several means")

ax2 = plt.subplot(2, 1 ,2)
plt.xlim(-plot_x_range, plot_x_range)
plt.ylim(0,plot_y_range)
plt.title("Normal distribution with several variances")

for shifted_u in range(-2,3):
    n_y = [f_X(x, shifted_u, v) for x in X]
    ax1.plot(X, n_y, label='u='+str(shifted_u))
    
for different_v in [(8+10*i)/10 for i in range(5)]:
    n_y = [f_X(x, u, different_v) for x in X]
    ax2.plot(X, n_y, label='v='+str(different_v))
    
ax1.legend(loc="best")
ax2.legend(loc="best")
plt.show()
```
