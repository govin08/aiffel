# 1. 대푯값과 평균

다섯 개의 숫자 $1,\:2,\:3,\:4,\:5\tag{a}$
의 **평균**은 얼마일까?

평균은
$$m=\frac{1+2+3+4+5}5=3$$
와 같이 계산할 수 있다.
이때 평균은, 위의 다섯 개의 숫자를 대표하는 값이라고 생각할 수 있다.
즉
"1,2,3,4,5는 평균적으로 3이다"라고 말할 수 있다.
또, 만약 월요일에는 과제를 한 개 끝냈고, 화요일에는 두 개 끝냈으며, 수요일에는 세 개, 목요일에는 네 개, 금요일에는 다섯 개 끝냈다면, 평균적으로 하루 당 과제를 세 개 끝냈다고 말할 수도 있다.

이와 같이, 주어진 자료를 대표하는 값을 **대푯값**이라고 한다.
대푯값에는 다음 세 가지가 가장 대표적이다.
 - 평균 (mean, average)
 - 중앙값 (중위값, median)
 - 최빈값 (mode)
 
하지만 이 글에서는 '평균'만을 생각하려고 한다.
(또한 평균이라는 개념에도 산술평균과 기하평균, 조화평균, 가중치를 고려한 평균 등 종류가 굉장히 많지만, 산술평균에 대해서만 이야기한다.)

(a)에서 적었던 다섯 개의 숫자는 **변량**이라고 부른다.
일반적으로, 변량 $x_1, x_2, x_3,\cdots,x_n$에 대하여 이 변량들의 평균 $m$은
$$m = \frac{x_1+x+2+\cdots+x_n}n$$
이다.
이것을 합의 기호 $\Sigma$를 사용해 표현하면
$$m = \frac1n\sum_{i=1}^nx_i$$
이라고 적을 수도 있을 것이다.

# 2. 산포도

산포도란, 변량들이 대푯값으로부터 얼만큼이나 떨어져있느냐 하는 정도를 나타내는 값이다.
우리는 대푯값을 평균만 고려하기로 했으므로 산포도란
"변량들이 평균으로부터 얼마나 떨어져 있는지"
를 나타내는 값이다.

예를 들어, (a) 1, 2, 3, 4, 5와 (d) 1, 2, 4, 5, 8를 생각해보자.
이것들을 수직선 위에 표시하면 다음과 같다.

![](https://i.imgur.com/9s7LbKc.png)

![](https://i.imgur.com/GkpaG0h.png)

둘 중에서 어떤 점들이 더 모여있고 어떤 점들이 더 퍼져있을까?
상대적으로 (a)의 점들이 조금 더 모여있고, (b)의 점들은 멀리 퍼져있는 것을 알 수 있다.
조금 더 정확하게는, (a)의 점들은 평균 3을 중심으로 모여있는 편이고, (b)의 점들은 평균 4를 중심으로 조금 떨어져서 분포한다고 말할 수 있다.
그러니까, (b)의 산포도가 (a)의 산포도보다 더 크다고 말할 수 잇을 것이다.

그런데 "변량들이 평균으로부터 얼마나 떨어져 있는지"를 고려하려면, 단순히 변량에서 평균을 뺀 값을 생각하면 될 것이다.
이처럼 변량에서 평균을 뺀 값을 '편차'라고 한다.

<center>편차 = 변량 - 평균</center>

각각의 변량에 대하여 편차를 계산하면 다음과 같다.

|변량|1 |2 |3|4|5|
|:-:|:-:|:-:|:-:|:-:|:-:|
|편차|-2|-1|0|1|2|

산포도란 "변량들이 평균으로부터 얼마나 떨어져있는지"를 평균적으로 계산한 값이다.
그러니까, 편차에 대해서 적당히 평균을 내면 그것이 산포도의 역할을 할 수 있을 것이다.
그런데 편차에 대하여 평균을 계산하면 0이 나온다.

$$\frac{(-2)+(-1)+0+1+2}5=0$$

이것은 이상한 일이 아니다.
어떤 종류의 자료라고 하더라도, 편차의 평균은 항상 0이다.
만약 변량들이
$$x_1,\:x_2,\:\cdots,\:x_n$$
이라고 하면 각 변량에 대응되는 편차들은
$$x_1-m,\:x_2-m,\:\cdots,\:x_n-m$$
이다.
이것들의 평균을 계산하면
\begin{align*}
\text{편차들의 평균}
&=\frac{(x_1-m)+(x_2-m)+\cdots+(x_n-m)}n\\
&=\frac{(x_1+x_2+\cdots+x_n)-(m+m+\cdots+m)}n\\
&=\frac{x_1+x_2+\cdots+x_n}n-\frac{mn}n\\
&=m-m=0
\end{align*}
이다.
그러니까, 정말로 편차의 평균은 항상 0이 된다.

따라서 편차의 평균은 산포도로서의 의미를 가질 수가 없다.
다른 방식의 산포도를 고려해야 한다.

그런데 잘 생각해보면, 편차들의 평균이 0이 되었던 이유는, 편차들이 어떤 값은 양수(+)이고 또 어떤 값은 음수(-)였기 때문에 그렇다.
그것들을 다 더했더니 0이 나왔던 것이다.
그러면,

(1) 편차들을 제곱해서 그것들의 평균을 내는 게 한가지 방법이 될 수 있을 것이다.
원래 값이 양수였건 음수였건 간에 제곱을 하게 되면 항상 양수가 되니까 이것은 산포도로서의 의미를 가질 수 있다.
이렇게 얻는 산포도를 **분산(deviation)**이라고 한다.

(2) 그런데 분산은, 원래의 값에 제곱을 했기 때문에, 완전한 의미에서의 평균이라고는 할 수 없다.
그래서 분산에 근호(루트)를 씌우면 조금 더 정확한 의미의 산포도라고 할 수 있다.
이렇게 얻는 산포도를 **표준편차(standard deviation)**이라고 한다.

(3) 편차들에 절댓값을 씌워서 그것들의 평균을 내는 것도 방법이 될 수 있다.
이렇게 하여도 산포도의 역할을 수행할 수 있을 것이다.
이렇게 얻은 산포도는 **평균절대편차(mean absolute deviation)** 라고 한다.

이상을 정리하면 다음과 같다.

\begin{align*}
0\:=\:&\text{편차의 평균}\\
\text{분산}\:=\:&\text{[편차]}^2\text{의 평균}\\
\text{표준편차}\:=\:&\sqrt{\text{[편차]}^2\text{의 평균}}\\
\text{평균절대편차}\:=\:&|편차|\text{의 평균}\\
\end{align*}

분산을 $V$로, 표준편차를 $\sigma$로 평균절대편차를 $\text{MAD}$라고 적자.
그러면, 변량 $x_1$, $x_2$, $\cdots$, $x_n$에 대하여, 분산과 표준편차와 평균절대편차를 구하는 식은 다음과 같이 될 것이다.
\begin{align*}
V
&=\frac{(x_1-m)^2+(x_2-m)^2+\cdots+(x_n-m)^2}n=\frac1n\sum_{i=1}^n(x_i-m)^2\\
\sigma &= \sqrt V\\
\text{MAD}
&=\frac{|x_1-m|+|x_2-m|+\cdots+|x_n-m|}n=\frac1n\sum_{i=1}^n|x_i-m|
\end{align*}

예를 들어 (a) 1, 2, 3, 4, 5의 분산과 표준편차, 평균절대편차는 얼마일까?
먼저 분산은
$$V = \frac{(1-3)^2+(2-3)^2+(3-3)^2+(4-3)^2+(5-3)^2}5=2$$
이다.
따라서 표준편차는
$$\sigma = \sqrt V = \sqrt2$$
이고, 평균절대편차는
$$\text{MAD}=\frac{|1-3|+|2-3|+|3-3|+|4-3|+|5-3|}5=\frac65$$
이다.

# 3. MSE, RMSE, MAE

## 3.1. MSE, RMSE, MAE
회귀분석(Regression analysis)은, 관측된 데이터들에 대하여, 그 데이터들을 가장 잘 만족하게 하는 함수(모델)를 찾는 일을 말한다.
(이때 데이터란, 독립변수 $x$와 종속변수 $y$로 이루어진 순서쌍 $(x,y)$들의 집합이다.)
즉 함수(모델)로 예측한 예측값($\hat y_i$)이 관측값($y_i$)과 적게 차이나도록 함수(모델)을 구성하도록 한다.

예측값($\hat y_i$)과 관측값($\hat y$) 사이의 차이를 오차(error)라고 하면, 이 오차는 데이터의 개수만큼 존재한다.
즉 데이터가 $n$개 존재한다면, 오차는 $n$개의 실수 
$$\epsilon_1, \epsilon_2, \cdots, \epsilon_n$$
라고 생각할 수 있다.

우리는, 하나의 데이터에 대하여 오차가 얼마인지를 알고 싶어하는 것이 아니라 전체적으로 오차가 얼마나 되는지를 알고 싶어한다.
즉, 평균적으로 오차가 얼마나 발생했는지 알 수 있는 지표가 있어야 한다.
이러한 지표는 대표적으로 세 가지가 있다.
 - MSE : Mean Squared Error
 - RMSE : Root Mean Squared Error
 - MAE : Mean Absolute Error
 
MSE는 오차를 제곱(square)한 후 평균(mean)낸 것을 말하고, RMSE는 MSE에 근호(root)를 씌운 것을 말한다.
또한 MAE는 오차에 절댓값(absolute value)을 취한 뒤 평균(mean)낸 것을 말한다.
따라서 정확한 식은 다음과 같다.

\begin{align*}
\text{MSE}
&=\frac{{\epsilon_1}^2+{\epsilon_2}^2+\cdots+{\epsilon_n}^2}n\\
\text{RMSE}
&= \sqrt{\text{MSE}}\\
\text{MAE}
&=\frac{|\epsilon_1|+|\epsilon_2|+\cdots+|\epsilon_n|}n
\end{align*}

회귀분석, 특히 머신러닝과 관련된 회귀분석에서는 이러한 지표들을 고려하는 것이 상당히 중요하다.
특히 MSE는 대부분의 머신러닝 문제에서 cost function의 역할을 수행한다.

## 3.2. 산포도와의 관계
MSE, RMSE, MAE의 정의를 보면, 분산, 표준편차, 평균절대편차의 정의와 상당히 유사하다는 것을 관찰할 수 있다.
실제로, MSE, RMSE, MAE의 정의에서 '오차'라는 말을 '편차'라는 말로 바꾸면, 정확하게 분산, 표준편차, 평균절대편차의 정의와 일치한다.

## 3.3. L1, L2 norm / 정규화(Normalization)
$n$차원 벡터 $v=(v_1,v_2,\cdots,v_n)$에 대하여 이 벡터의 크기(norm)를 정하는 방법은 여러가지가 있다.
가장 표준적인 방법은 다음 식을 이용하는 것이다.
$$||v||_2 = \sqrt{{v_1}^2+{v_2}^2+\cdots+{v_n}^2}$$
이것은, 고등학교 수학에서 배우는 이차원 벡터 혹은 삼차원 벡터의 크기를 정의하는 방법과 일치한다.
한편, 다음과 같이 정의하여도 여전히 크기(norm)의 의미를 가진다.
$$||v||_1 = |v_1|+|v_2|+\cdots+|v_n|$$
이때, 앞의 식에서 정의된 norm을 L2 norm이라고 부르고, 뒤의 식에서 정의된 norm을 L1 norm이라고 부른다.
일반적으로는 실수 $p\ge1$에 대하여 Lp norm을 정의할 수도 있다.

위의 회귀분석에 대한 설명에서 오차란 $n$개의 실수값이라고 했었다.
즉, 특정한 모델에 대하여 오차는 다음과 같은 벡터라고 간주할 수 있다.

$$\epsilon = (\epsilon_1,\epsilon_2,\cdots,\epsilon_n)$$

그러면 RMSE는 벡터 $\epsilon$의 L2 norm과 관련이 있고, MAE는 벡터 $\epsilon$의 L1 norm과 관련이 있을 것이다.
정확하게는

$$RMSE = \frac{||\epsilon||_2}n$$

$$MAE = \frac{||\epsilon||_1}n$$

의 관계가 성립할 것이다.

한편, 머신러닝 및 딥러닝에서는 과적합(overfitting)을 막기 위한 방법 중 하나로 정규화(Regularization)라는 것이 있다.
여기에서도 L1 norm과 L2 norm이 사용된다.
L1 norm을 사용하는 방식을 Lasso방식이라고 하고, L2 norm을 사용하는 방식을 Ridge 방식이라고 한다.
