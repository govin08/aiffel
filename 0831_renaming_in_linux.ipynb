# 리눅스에서 파일 이름/파일 확장자 바꾸기

리눅스에서는 파일 이름을 바꾸는 것이 그렇게까지 쉬운 작업은 아닌 것 같습니다.
`rename`이라는 명령어가 기본적으로 제공되기는 하지만, 직접 써보니 오류가 너무 많이 납니다.
파일 이름 변경을 위해서는 다음 둘 중 하나를 선택해야 하는 것 같습니다.
 - `rename` 명령어를 고수하되 `regex`라고 하는 문법에 익숙해져야 한다.
 - 적당한 package를 다운받아서 그걸로 쓴다.
 
저는 후자를 선택했습니다.
여기서 제가 말하는 적당한 package란 `rename.ul`이라는 package입니다.

처음 상황은 다음과 같습니다.

현재 위치는 `test`라는 디렉토리인데, 이 디렉토리는 빈 디렉토리입니다.
이 `test`디렉토리 내에 세 개의 `.ipynb` 파일을 만들었습니다.
 - `apple1.ipynb`
 - `apple2.ipynb`
 - `apple3.ipynb`



이 상황에서 바로 `rename` 명령어를 사용해보면



와 같이 뜹니다.
뭔가 문법을 안지켰다는 것이고, 구글링을 해보면 `regex`라는 문법에 맞춰서 잘 쓰면 된다는데 너무 어렵습니다.
그래서 이 방법은 빠르게 포기했어요.

대신 다음과 같이 합니다.
먼저 `rename.ul`와 관련된 패키지를 설치합니다.
하지만
- `sudo apt install rename.ul`

와 같이 입력해서는 설치가 되지 않는 것 같습니다.
그런 이름을 가진 패키지가 없다고 합니다.



대신
- `sudo apt install util-linux`

라고 입력하여 `util-linux`라는 패키지를 깔면 `rename.ul`를 사용할 수 있게 됩니다.
[이곳](https://command-not-found.com/rename.ul)에서 해당 명령어를 가져왔어요.

이 `rename.ul`를 사용해 파일 이름을 바꿀 수 있고, 파일의 확장자도 바꿀 수 있습니다.
먼저 파일들의 이름을
 - `grape1.ipynb`
 - `grape2.ipynb`
 - `grape3.ipynb`

로 일괄적으로 바꾸는 명령어는
 - `rename.ul apple grape *`
 
입니다. `rename.ul`은 세 개의 인자 `apple`, `grape`, `*`를 받고 있습니다.
'apple'이라는 문자열을 모두 'grape'로 바꾸라는 뜻입니다.
마지막 인자 `*`sms '모든 파일에 대하여 시행하라'는 뜻입니다.

이 파일들의 확장자를
 - `grape1.py`
 - `grape2.py`
 - `grape3.py`

로 일괄적으로 바꾸는 명령어는
 - `rename.ul ipynb py *`

입니다.
아래 캡쳐를 보면 잘 바뀌었다는 것을 확인할 수 있습니다.
