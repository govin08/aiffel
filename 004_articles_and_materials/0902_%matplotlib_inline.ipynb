```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
x = np.linspace(0,5*np.pi,1000)
plt.plot(x, np.sin(x**2))
plt.title('A simple chirp')
#plt.show
```

아래 코드에서, 소위 magic method로 불리는

`%matplotlib inline`

은 주석처리를 하나 안하나 같은 결과가 나온다.
인터넷에 저 명령어에 대한 설명이 많이 나와있지만, 정작 저 명령어를 넣지 않는다고 해도 같은 결과가 나오니, 도무지 어떤 역할을 하는지 알 수가 없다.

그렇게 자료들을 열심히 찾아보다가 구글에 "without matplotlib inline"라고 검색했더니 stack overflow에
[다음과 같은 글](https://stackoverflow.com/questions/65934740/is-matplotlib-inline-still-needed)
이 있다.

이 질문자도 나와 같은 의문을 품었다.
맨 첫문단밖에 읽지는 않긴 했다.
근데 어쨌든 맨 첫문단의 내용은, 바로 내가 위에 적었던 내용과 거의 일치하는 것이다.

아래와 같은 글이다.

이에 대한 첫번째 답변은 무슨 소린지 잘 이해가 가지 않는다.
하지만 두번째 답변은 알아들을 수 있을 것 같아 아래에 적어보았다.

요약하면, `%matplotlib inline` 명령어는 아주 특별한 경우가 아니면, 더이상 사용될 필요가 없다는 것이다.
그것은 아마도 ipython이나 jupyter notebook의 버전이 계속 계정됨에 따라, 굳이 해당 명령어를 넣지 않아도 되게끔 업데이트되었기 때문이라는 것 같다.
하지만, 관습(convention)상 계속 `%matplotlib inline`를 써주는 것이 안전하다는 것으로 이해했다.
