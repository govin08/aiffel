## 1.1. 도입

 - 캐글 데이터셋 instacart market basket analysis는 2017년에 처음으로 공개된 데이터셋이다.
 - 300만 개의 식료품 주문(grocery orders)과 20만개의 소비자(users)에 대한 데이터가 담겨있다.
 - 각각의 소비자들은 4개에서 100개 정도의 주문을 수행하였다.
 - 각각의 주문들에 대하여 다음과 같은 feature들이 표시되었다.
   - 무슨 요일에 주문했는지
   - 몇시에 주문했는지
   - 이전 주문과 비교하여 며칠의 날짜 간격이 있는지.
 - 이 데이터셋에 대하여 다음 두 가지의 결론을 간단하게 도출해보았다.
   - 데이터들이 나타나는 패턴
   - 저녁 시간대에 가장 많이 팔린 제품은 무엇인가?

 - 개인정보를 잘 고려하여 법에 저촉되지 않게 데이터셋을 구성하였다는 설명이 있다.

 여러 사람에게서 구매되었고, 여러 판매자에게서 구매되는 경우의 상품만이 데이터셋에 포함되어 있다.
 (원문 :  Only products that are bought by multiple people at multiple retailers are included, and no retailer ID is provided.)


## 1.5. Some interesting findings

### 1.5.1. 구매횟수와 재구매비율 사이의 관계

위의 그림은, `aisle`들을 xy평면에 시각화한 것이다.
각각의 점들은 aisle들을 나타내는데, 점들에 `aisle`이 표시된 것이 아니라, 해당 `aisle`이 속해 있는 `department`가 표시된 것으로 보인다.

이때 x축은 구매횟수를 나타내고, y축은 재구매비율을 나타낸다.
구매 횟수와 재구매비율은 양의 상관관계를 가지는 것이 보인다.
(원문 : For example, this plot shows that the most common aisles are more likely to be reordered from.)

과일은 채소보다 자주 재주문된다.
이것은 채소는 과일에 비해 간헐적으로(intermittently) 쓰이기 때문이다. [이게 무슨 말인지 잘 모르겠다.] (원문 : Fruits are reordered more frequently than vegetables — perhaps because vegetables are more intermittently purchased for recipes.)

수프와 재빵 재료같은 주식(主食)들은 재주문되는 빈도가 적다.
(원문 : Staples like soups and baking ingredients are least likely to be reordered — perhaps because they are less frequently needed.)
