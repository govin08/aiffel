# 3. 해석

정확히 말하면, `(X_test[:, 0], y_test)`는 실제 데이터라고 말하는 것과 `(X_test[:, 0], prediction)`를 예측 데이터라고 말하는 데에는 조금 무리가 있습니다.
`X_test`가 아니라 `X_test[:,0]`을 고려하고 있기 때문입니다.
다시 말해,
아홉 개의 독립변수 $x_0,\cdots,x_9$들 중 첫번째 독립변수인 $x_0$만 고려하고 있다는 말입니다.

`prediction`과 `X_test`($=x_0,x_1,\cdots,x_9$)는 선형적인 관계에 있는 것이 맞습니다.
즉

$$\text{prediction}=w_0x_0+w_1x_w+\cdots+w_9x_9+b$$

의 관계식을 만족시킵니다.
(그래서, 11차원 공간 위에 놓여 있는 위 식의 그래프인 10차원 평면을 생각하면, 주황색 점에 해당하는 점들은, 그 평면 위에 놓여있을 겁니다.
마치 첫번째 그래프에서 주황색 점들이 한 직선 위에 놓여있듯이요.)

하지만, `prediction`과 `X_test[:,0]`($=x_0$)는 선형적인 관계를 만족시킬 필요가 없습니다.
즉,

$$\text{prediction}=w'x_0+b'$$

의 관계식을 만족시킬 필요는 없습니다.
따라서, 두번재 그래프에서 주황색 점들이 하나의 선 위에 나타나지 않는 것이 정상입니다.
