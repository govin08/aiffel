```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles, venn3, venn3_circles
```

# LMS fundamental 15

## 15.5. 집합 실행해보기

```python
A = 3 # n(A-B)
B = 2 # n(B-A)
AB = 1 # n(A\cap B)
```

### 15.5.1. 기본 벤다이어그램

```python
v = venn2(subsets={'10': A, '01': B, '11': AB}, set_labels = ('A', 'B')) 
c = venn2_circles(subsets=(A, B, AB), linestyle='solid') 
v.get_label_by_id('10').set_text('') 
v.get_label_by_id('01').set_text('') 
v.get_label_by_id('11').set_text('')

plt.gca().set_facecolor("white")
plt.gca().set_axis_on()
```

### 15.5.2. 교집합

```python
v = venn2(subsets={'10': A, '01': B, '11': AB}, set_labels = ('A', 'B')) 
c = venn2_circles(subsets=(A, B, AB), linestyle='solid') 

v.get_patch_by_id('10').set_color('w') 
v.get_patch_by_id('01').set_color('w') 
v.get_patch_by_id('11').set_color('gold') 

v.get_label_by_id('10').set_text('') 
v.get_label_by_id('01').set_text('') 
v.get_label_by_id('11').set_text('')

plt.gca().set_facecolor("white")
plt.gca().set_axis_on()
```

### 15.5.3. 합집합

```python
v = venn2(subsets={'10': A, '01': B, '11': AB}, set_labels = ('A', 'B')) 
c = venn2_circles(subsets=(A, B, AB), linestyle='solid') 

v.get_patch_by_id('10').set_color('gold') 
v.get_patch_by_id('01').set_color('gold') 
v.get_patch_by_id('11').set_color('gold') 

v.get_label_by_id('10').set_text('') 
v.get_label_by_id('01').set_text('') 
v.get_label_by_id('11').set_text('')

plt.gca().set_facecolor("white")
plt.gca().set_axis_on()
```

### 15.5.4. 차집합

```python
v = venn2(subsets={'10': A, '01': B, '11': AB}, set_labels = ('A', 'B')) 
c = venn2_circles(subsets=(A, B, AB), linestyle='solid') 

v.get_patch_by_id('10').set_color('gold') 
v.get_patch_by_id('01').set_color('white') 
v.get_patch_by_id('11').set_color('white') 

v.get_label_by_id('10').set_text('') 
v.get_label_by_id('01').set_text('') 
v.get_label_by_id('11').set_text('')

plt.gca().set_facecolor("white")
plt.gca().set_axis_on()
```

### 15.5.5. 여집합

```python
v = venn2(subsets={'10': A, '01': B, '11': AB}, set_labels = ('A', 'B')) 
c = venn2_circles(subsets=(A, B, AB), linestyle='solid') 

v.get_patch_by_id('10').set_alpha(1)
v.get_patch_by_id('01').set_alpha(1)
v.get_patch_by_id('11').set_alpha(1)

v.get_patch_by_id('10').set_color('white')
v.get_patch_by_id('01').set_color('khaki')
v.get_patch_by_id('11').set_color('white')

v.get_label_by_id('10').set_text('') 
v.get_label_by_id('01').set_text('') 
v.get_label_by_id('11').set_text('')

plt.gca().set_facecolor("khaki")
plt.gca().set_axis_on()
```

### 15.5.6. 서로소인 집합

```python
v = venn2(subsets={'10': A, '01': B}, set_labels = ('A', 'B')) 
c = venn2_circles(subsets=(A, B, 0), linestyle='solid') 

v.get_patch_by_id('10')
v.get_patch_by_id('01')

v.get_patch_by_id('10').set_color('gold')
v.get_patch_by_id('01').set_color('peachpuff')

v.get_label_by_id('10').set_text('') 
v.get_label_by_id('01').set_text('')

plt.gca().set_facecolor("white")
plt.gca().set_axis_on()
```

```python
v = venn2(subsets={'10': A, '01': B}, set_labels = ('Set A', 'Set B'))
c = venn2_circles(subsets=(A, B, 0), linestyle='solid')

v.get_patch_by_id('10').set_alpha(0.5)
v.get_patch_by_id('01').set_alpha(0.5)
v.get_patch_by_id('10').set_color('r')
v.get_patch_by_id('01').set_color('gray')
v.get_label_by_id('10').set_text('')
v.get_label_by_id('01').set_text('')
v.get_label_by_id('A').set_text('A')
v.get_label_by_id('B').set_text('B')

plt.gca().set_axis_on()
```
